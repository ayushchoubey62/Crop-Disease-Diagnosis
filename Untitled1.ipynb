```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define training and validation data generators
train_datagen = ImageDataGenerator(
    rescale=1.0/255, 
    rotation_range=20, 
    width_shift_range=0.2, 
    height_shift_range=0.2, 
    shear_range=0.2, 
    zoom_range=0.2, 
    horizontal_flip=True,
    validation_split=0.2  # 20% data for validation
)

# Load training and validation data
train_generator = train_datagen.flow_from_directory(
    r'C:\Users\ayush\OneDrive\Desktop\project 1\New Plant Diseases Dataset(Augmented)\New Plant Diseases Dataset(Augmented)\train',  # Replace with the correct dataset path
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    r'C:\Users\ayush\OneDrive\Desktop\project 1\New Plant Diseases Dataset(Augmented)\New Plant Diseases Dataset(Augmented)\valid',
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, Input

model = models.Sequential([
    Input(shape=(150, 150, 3)),  # Define input explicitly
    layers.Conv2D(32, (3, 3), activation="relu"),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3, 3), activation="relu"),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(128, (3, 3), activation="relu"),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(128, (3, 3), activation="relu"),
    layers.MaxPooling2D(2, 2),

    layers.Flatten(),
    layers.Dense(512, activation="relu"),
    layers.Dense(38, activation="softmax")
])

model.summary()
```

```python
model.compile(
    optimizer="adam",  
    loss="categorical_crossentropy",  # Use "sparse_categorical_crossentropy" if labels are not one-hot encoded
    metrics=["accuracy"]
)

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=5,  # Increase epochs if needed
    verbose=1
)
```

```python
model.save('my_model.keras')
```

```python
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog, messagebox, Toplevel
from PIL import Image, ImageTk, ImageEnhance, ImageFilter, ImageOps, ImageGrab
import time
import threading
import os
from fpdf import FPDF
import numpy as np
import cv2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import sys
import logging
from datetime import datetime
import json
import shutil
import tempfile
import webbrowser # Added for opening images/PDFs
import re # Added for email validation

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s', filename='crop_diagnosis.log')

class App(tk.Tk):
    """
    Main application class for the Crop Disease Diagnosis App.
    Handles overall application flow, theme management, model loading,
    and data persistence.
    """
    def __init__(self):
        super().__init__()

        # Hide the main window initially to show splash screen
        self.withdraw()

        # Initialize core application data and models
        self.disease_model = None
        self.crop_classifier = None # Placeholder for a potential crop classifier model
        self.class_labels = None
        self.recent_images = []
        self.diagnosis_history = []
        
        self.current_image_path = None
        self.result_frame = ttk.Frame(self) # This might be a leftover, typically frames are managed by show_frame

        self.frames = {} # Dictionary to hold instances of different pages (frames)
        
        # Initialize theme state and color dictionary
        self.dark_mode = False # Default to light mode
        self.colors = {} # Will be populated by _apply_theme

        # Initialize notification label (will be created dynamically)
        self.notification_label = None
        self.notification_timer = None # To manage notification display duration

        # Set up window closing protocol
        self.protocol("WM_DELETE_WINDOW", self.quit)

        # Display splash screen and then load app components in a separate thread
        self.show_splash_screen()

    def show_splash_screen(self):
        """Displays a splash screen while the app loads resources."""
        self.splash_root = Toplevel(self)
        self.splash_root.overrideredirect(True) # Remove window decorations for a cleaner splash

        # Center the splash screen on the monitor
        screen_width = self.splash_root.winfo_screenwidth()
        screen_height = self.splash_root.winfo_screenheight()
        splash_width = 400
        splash_height = 280 # Increased height for progress bar
        x = (screen_width // 2) - (splash_width // 2)
        y = (screen_height // 2) - (splash_height // 2)
        self.splash_root.geometry(f"{splash_width}x{splash_height}+{x}+{y}")
        
        # Apply a simple theme to the splash screen
        splash_bg_color = "#34495E" # Dark blue-grey
        splash_fg_color = "#ECF0F1" # Light grey
        self.splash_root.config(bg=splash_bg_color)

        # Splash image/logo
        logo_path = None
        try:
            # Attempt to load the logo image
            logo_path = self.get_resource_path(os.path.join("assets", r"C:\Users\ayush\OneDrive\Desktop\project 1\CropDiseaseApp\assets\icons8-chaff-48.png"))
            splash_img = Image.open(logo_path).resize((150, 150), Image.Resampling.LANCZOS)
            self.splash_photo = ImageTk.PhotoImage(splash_img)
            logo_label = tk.Label(self.splash_root, image=self.splash_photo, bg=splash_bg_color)
            logo_label.pack(pady=10) # Adjusted padding
        except FileNotFoundError:
            logging.warning(f"Splash screen logo not found at {logo_path}. Using text fallback.")
            logo_label = tk.Label(self.splash_root, text="Crop Diagnosis", font=("Helvetica", 24, "bold"), fg=splash_fg_color, bg=splash_bg_color)
            logo_label.pack(pady=10)
        except Exception as e:
            logging.error(f"Error loading splash screen logo from {logo_path}: {e}")
            logo_label = tk.Label(self.splash_root, text="Crop Diagnosis", font=("Helvetica", 24, "bold"), fg=splash_fg_color, bg=splash_bg_color)
            logo_label.pack(pady=10)

        # Loading message (dynamic)
        self.splash_loading_message = tk.StringVar()
        self.splash_loading_message.set("Initializing...")
        loading_label = tk.Label(self.splash_root, textvariable=self.splash_loading_message, font=("Helvetica", 12), fg=splash_fg_color, bg=splash_bg_color)
        loading_label.pack(pady=5)

        # Progress bar
        self.splash_progress = ttk.Progressbar(
            self.splash_root,
            orient="horizontal",
            length=300,
            mode="determinate"
        )
        self.splash_progress.pack(pady=10)

        # Start loading main app components in a separate thread to avoid freezing UI
        threading.Thread(target=self._load_app_components_and_show_main, daemon=True).start()

    def _load_app_components_and_show_main(self):
        """
        Loads models and application data, then shows the main application window.
        This runs in a separate thread.
        """
        try:
            # 1. Load application data (history and theme preference)
            self.after(0, lambda: self.splash_loading_message.set("Loading application data..."))
            self.after(0, lambda: self.splash_progress.config(value=25)) # Corrected: use config(value=X)
            self._load_app_data()
            logging.info("Application data loaded.")

            # 2. Apply theme based on loaded preference *before* creating UI elements
            self.after(0, lambda: self._apply_theme(self.dark_mode))
            logging.info(f"Initial theme applied: {'Dark' if self.dark_mode else 'Light'} Mode. (from _load_app_components_and_show_main)")

            # 3. Load machine learning models
            self.after(0, lambda: self.splash_loading_message.set("Loading machine learning models..."))
            self.after(0, lambda: self.splash_progress.config(value=50)) # Corrected: use config(value=X)
            self.load_models()
            # Define class labels - make sure these match your model's output
            # This is crucial for interpreting model predictions.
            self.class_labels = {
                'Apple_Apple_scab': 0, 'Apple_Black_rot': 1, 'Apple_Cedar_apple_rust': 2, 
                'Apple_healthy': 3, 'Blueberry_healthy': 4, 'Cherry_Powdery_mildew': 5, 
                'Cherry_healthy': 6, 'Corn_Cercospora_leaf_spot_Gray_leaf_spot': 7, 
                'Corn_Common_rust': 8, 'Corn_Northern_Leaf_Blight': 9, 'Corn_healthy': 10, 
                'Grape_Black_rot': 11, 'Grape_Esca_Black_Measles': 12, 'Grape_Leaf_blight': 13, 
                'Grape_healthy': 14, 'Orange_Haunglongbing_Citrus_greening': 15, 
                'Peach_Bacterial_spot': 16, 'Peach_healthy': 17, 'Pepper_bell_Bacterial_spot': 18, 
                'Pepper_bell_healthy': 19, 'Potato_Early_blight': 20, 'Potato_Late_blight': 21, 
                'Potato_healthy': 22, 'Raspberry_healthy': 23, 'Soybean_healthy': 24, 
                'Squash_Powdery_mildew': 25, 'Strawberry_Leaf_scorch': 26, 'Strawberry_healthy': 27, 
                'Tomato_Bacterial_spot': 28, 'Tomato_Early_blight': 29, 'Tomato_Late_blight': 30, 
                'Tomato_Leaf_Mold': 31, 'Tomato_Septoria_leaf_spot': 32, 
                'Tomato_Spider_mites _Two_spotted_spider_mite': 33, 'Tomato_Target_Spot': 34, 
                'Tomato_Tomato_Yellow_Leaf_Curl_Virus': 35, 'Tomato_Tomato_mosaic_virus': 36, 
                'Tomato_healthy': 37
            }
            logging.info("Class labels defined.")

            # 4. Initialize main UI components on the main thread
            self.after(0, lambda: self.splash_loading_message.set("Setting up user interface..."))
            self.after(0, lambda: self.splash_progress.config(value=75)) # Corrected: use config(value=X)
            self.after(0, self.setup_main_app)
            logging.info("Main application UI setup initiated.")

        except Exception as e:
            logging.critical(f"Fatal error during application startup: {e}", exc_info=True)
            # Show a critical error message and then exit
            self.after(0, lambda: messagebox.showerror("Fatal Error", f"The application failed to start due to a critical error:\n{str(e)}\n\nPlease check the logs for more details and restart the application."))
            self.after(100, self.quit) # Schedule quitting after error message is displayed
        finally:
            # Always destroy splash screen and show main window (or attempt to)
            self.after(0, lambda: self.splash_loading_message.set("Starting application..."))
            self.after(0, lambda: self.splash_progress.config(value=100)) # Corrected: use config(value=X)
            self.after(500, self._fade_out_splash) # Fade out splash screen
            self.after(1000, self.deiconify) # Show main window after fade out

    def _fade_out_splash(self):
        """Fades out the splash screen over a short duration."""
        alpha = self.splash_root.attributes("-alpha")
        if alpha > 0:
            alpha -= .1 # Decrease alpha
            self.splash_root.attributes("-alpha", alpha)
            self.splash_root.after(50, self._fade_out_splash)
        else:
            self.splash_root.destroy() # Destroy splash screen when fully transparent


    def setup_main_app(self):
        """
        Initializes the main application UI components.
        This method is called on the main thread after loading is complete.
        """
        self.title("Crop Disease Detection")
        self.geometry("1000x700")
        self.minsize(800, 600)
        
        # Create base styles. Specific widget styles are configured in _configure_widget_styles.
        self.create_styles() 

        # Create container for all frames (pages)
        container = tk.Frame(self)
        container.pack(fill="both", expand=True)
        container.grid_rowconfigure(0, weight=1)  
        container.grid_columnconfigure(0, weight=1)

        # Import page classes locally to avoid circular dependencies during initial setup
        from __main__ import HomePage, DiagnosisPage, HistoryPage, FeedbackPage, HelpPage, AboutPage

        # Create instances of all pages and store them in the frames dictionary
        for F in (HomePage, DiagnosisPage, HistoryPage, FeedbackPage, HelpPage, AboutPage):
            page_name = F.__name__
            frame = F(parent=container, controller=self)
            self.frames[page_name] = frame
            frame.grid(row=0, column=0, sticky="nsew")

        # Show the initial page
        self.show_frame("HomePage")
        logging.info("Main application UI setup complete. Displaying HomePage.")

    def create_styles(self):
        """
        Initializes the ttk.Style object and defines base styles.
        Actual color mapping based on theme is done in _configure_widget_styles.
        """
        self.style = ttk.Style()
        
        # Default font for all widgets
        self.style.configure('.', font=('Inter', 12)) # Using 'Inter' as per instructions

        # Define basic styles that will be overridden by theme-specific colors
        self.style.configure('TFrame', borderwidth=0)
        self.style.configure('TLabel', borderwidth=0)
        self.style.configure('TLabelframe', borderwidth=1, relief="solid")
        self.style.configure('TLabelframe.Label', font=('Inter', 12, 'bold'))
        self.style.configure('TEntry', borderwidth=1, relief="solid")
        self.style.configure('TCombobox', borderwidth=1, relief="solid")
        self.style.configure('TProgressbar', borderwidth=0)

        # Specific styles that may have different border/relief
        self.style.configure('Card.TFrame',
                             borderwidth=2,
                             relief="raised", # Raised relief for a card effect
                             # lightcolor and darkcolor will be set by _configure_widget_styles
                             )
        
        self.style.configure('Toggle.TButton',
                             font=("Segoe UI Emoji", 12),
                             borderwidth=0,
                             relief='flat',
                             width=3)
        
        # Transparent styles for HomePage elements (used with canvas.create_window)
        self.style.configure('Transparent.TFrame', background='')
        self.style.configure('Transparent.TLabel', background='')


    def load_models(self):
        """Load machine learning models safely with robust error handling and warm-up."""
        model_files = {
            "disease_model": r"C:\Users\ayush\OneDrive\Desktop\project 1\CropDiseaseApp\models\my_model.keras"
            # "crop_classifier": "crop_classifier.keras"  # Uncomment if you have a crop classifier
        }

        for model_name, filename in model_files.items():
            try:
                # Common paths to check for model files
                possible_paths = [
                    os.path.join("models", filename),  # models/ folder in current directory
                    filename,  # just the file in the current directory
                    os.path.join(os.path.abspath(os.path.dirname(sys.argv[0])), "models", filename)  # safe base path for PyInstaller
                ]

                model_path = next((path for path in possible_paths if os.path.exists(path)), None)

                if not model_path:
                    error_msg = (f"Model file '{filename}' not found in expected locations. "
                                 "Please ensure it's in the 'models' folder or the application directory.")
                    logging.error(f"Failed to load {model_name}: {error_msg}")
                    # Raise a RuntimeError to stop app startup if a critical model is missing
                    raise RuntimeError(f"Missing critical model: {filename}. {error_msg}")

                logging.info(f"Loading {model_name} from: {model_path}")
                model = load_model(model_path)
                setattr(self, model_name, model)
                logging.info(f"{model_name} loaded successfully.")

                # Perform a warm-up prediction to avoid TensorFlow retracing warnings
                # This helps TensorFlow build and cache the computational graph for the expected input shape.
                if model_name == "disease_model":
                    try:
                        # Assuming target_size=(150, 150) for disease_model
                        dummy_input = np.zeros((1, 150, 150, 3), dtype=np.float32)
                        _ = model.predict(dummy_input, verbose=0) # verbose=0 to suppress output
                        logging.info(f"Warm-up prediction for {model_name} completed.")
                    except Exception as warm_up_e:
                        logging.warning(f"Failed to perform warm-up prediction for {model_name}: {warm_up_e}")

            except Exception as e:
                logging.critical(f"Failed to load {model_name}: {e}", exc_info=True)
                # Re-raise the exception to be caught by _load_app_components_and_show_main
                raise RuntimeError(f"Failed to load {model_name}: {str(e)}")

                    
    def get_resource_path(self, relative_path):
        """
        Get absolute path to resource, works for development and PyInstaller bundles.
        Prioritizes sys._MEIPASS for PyInstaller, then sys.argv[0] directory, then CWD.
        """
        # For PyInstaller bundled applications
        if hasattr(sys, '_MEIPASS'):
            base_path = sys._MEIPASS
            full_path = os.path.join(base_path, relative_path)
            if os.path.exists(full_path):
                logging.debug(f"Resource path resolved via _MEIPASS: {full_path}")
                return full_path

        # For development or non-bundled scripts (relative to script location)
        base_path = os.path.abspath(os.path.dirname(sys.argv[0]))
        full_path = os.path.join(base_path, relative_path)
        if os.path.exists(full_path):
            logging.debug(f"Resource path resolved via sys.argv[0] dir: {full_path}")
            return full_path

        # Fallback to current working directory
        current_working_dir = os.getcwd()
        full_path = os.path.join(current_working_dir, relative_path)
        if os.path.exists(full_path):
            logging.debug(f"Resource path resolved via CWD: {full_path}")
            return full_path
        
        # If none found, raise an error
        raise FileNotFoundError(f"Resource '{relative_path}' not found in any expected location.")


    def _apply_theme(self, is_dark_mode):
        """
        Applies the specified theme (dark or light) to the application.
        This method updates the color dictionary and reconfigures all widget styles.
        """
        self.dark_mode = is_dark_mode # Set the internal state

        # Define color schemes
        if self.dark_mode:
            self.colors = {
                'bg': '#1A1A1A',       # Dark background
                'fg': '#E0E0E0',       # Light text
                'primary': '#BB86FC',  # Purple accent
                'secondary': '#03DAC6',# Teal accent
                'surface': '#2C2C2C',  # Card surfaces, slightly lighter than bg
                'error': '#CF6679',    # Error color
                'textfield': '#3A3A3A',# Input fields background
                'button_bg': '#03DAC6',   # Teal button color (using secondary for high contrast)
                'button_hover_bg': '#00B8A9', # Slightly darker teal for hover
                'diagnosis_page_bg': '#263238', # Dark Blue Grey for DiagnosisPage
                'diagnosis_title_fg': '#E0E0E0', # Light text for title
                'diagnosis_title_hover_fg': '#03A9F4', # Light Blue for title hover
                'homepage_fg': 'white', # Text color for homepage in dark mode
                'homepage_button_hover_fg': '#03DAC6', # Teal for homepage button hover in dark mode
                'homepage_button_hover_bg': '#3A3A3A', # Slightly lighter dark for hover
                'image_canvas_bg': '#2C2C2C', # Dark background for image canvas
                'zoom_label_bg': '#2C2C2C', # Dark background for zoom label
                'help_page_title_fg': '#81C784', # Light green for help page title in dark mode
                'help_page_text_fg': '#A5D6A7', # Lighter green for help page text in dark mode
                'about_page_bg': '#263238', # Dark Blue Grey for AboutPage
                'about_page_title_fg': '#E0E0E0', # Light text for AboutPage title
                'about_page_text_fg': '#A5D6A7', # Lighter green for AboutPage text
                'border_color': '#555555', # General border color for dark mode
                'selected_item_bg': '#BB86FC', # Primary color for selected items (e.g., Treeview)
                'selected_item_fg': '#FFFFFF' # White text on selected items
            }
        else:
            self.colors = {
                'bg': '#F5F5F5',       # Light background
                'fg': '#212121',       # Dark text
                'primary': '#6200EE',  # Deep purple
                'secondary': '#03DAC6',# Teal
                'surface': '#FFFFFF',  # Card surfaces
                'error': '#B00020',    # Error color
                'textfield': '#FFFFFF',# Input fields
                'button_bg': '#0056B3',   # Button color
                'button_hover_bg': '#1976D2',     # Button hover
                'diagnosis_page_bg': '#E0F2F7', # Light Cyan for DiagnosisPage
                'diagnosis_title_fg': '#212121', # Dark text for title
                'diagnosis_title_hover_fg': '#673AB7', # Deep Purple for title hover
                'homepage_fg': 'black', # Text color for homepage in light mode
                'homepage_button_hover_fg': '#6200EE', # Deep purple for homepage button hover in light mode
                'homepage_button_hover_bg': '#E0E0E0', # Slightly darker light for hover
                'image_canvas_bg': 'white', # Light background for image canvas
                'zoom_label_bg': 'white', # Light background for zoom label
                'help_page_title_fg': '#2E7D32', # Dark green for help page title in light mode
                'help_page_text_fg': '#1B5E20', # Even darker green for help page text in light mode
                'about_page_bg': '#E0F2F7', # Light Cyan for AboutPage
                'about_page_title_fg': '#212121', # Dark text for AboutPage title
                'about_page_text_fg': '#1B5E20', # Even darker green for AboutPage text
                'border_color': '#CCCCCC', # General border color for light mode
                'selected_item_bg': '#6200EE', # Primary color for selected items (e.g., Treeview)
                'selected_item_fg': '#FFFFFF' # White text on selected items
            }
        
        # Configure all ttk widget styles with the new colors
        self._configure_widget_styles(self.colors)
        
        # Recursively update colors of all existing tk and ttk widgets
        self._update_widget_colors(self.colors)
        
        # Save theme preference for next launch
        self._save_theme_preference()
        
        # Provide visual feedback to the user
        self._show_notification(f"{'Dark' if self.dark_mode else 'Light'} Mode Enabled", "info")
        logging.info(f"Theme switched to {'Dark' if self.dark_mode else 'Light'} Mode.")


    def toggle_theme_mode(self):
        """Toggles the dark mode state and then applies the new theme."""
        new_dark_mode_state = not self.dark_mode
        self._apply_theme(new_dark_mode_state)


    def _configure_widget_styles(self, colors):
        """
        Configures the ttk widget styles with the provided color scheme.
        This method is called whenever the theme changes.
        """
        # Re-initialize style to ensure all previous maps are cleared and new ones are applied
        self.style = ttk.Style() 
        
        # General style for all widgets
        self.style.configure('.', 
                             font=('Inter', 12), # Use 'Inter' font
                             background=colors['bg'], 
                             foreground=colors['fg'],
                             fieldbackground=colors['textfield'])

        # Button styles
        self.style.configure('TButton',
                             background=colors['button_bg'],
                             foreground=colors['fg'], # Use theme's foreground color for contrast
                             bordercolor=colors['primary'],
                             lightcolor=colors['primary'],
                             darkcolor=colors['primary'],
                             padding=10,
                             relief='raised',
                             focusthickness=0, # Remove focus border
                             borderradius=5 # Rounded corners for buttons
                             )
        
        self.style.map('TButton',
                       background=[('active', colors['button_hover_bg']),
                                   ('disabled', colors['surface'])],
                       foreground=[('disabled', 'grey')]) # Keep disabled text grey for visibility

        # Hover button style (for specific buttons that need a different hover effect)
        self.style.configure('Hover.TButton',
                             background=colors['button_hover_bg'],
                             foreground=colors['fg'],
                             padding=10,
                             borderwidth=1,
                             relief='raised',
                             font=('Inter', 12, 'bold'))
        self.style.map('Hover.TButton',
                       background=[('active', colors['button_hover_bg'])])
        
        # Frame, Label, LabelFrame styles
        self.style.configure('TFrame', background=colors['bg'])
        self.style.configure('TLabel', background=colors['bg'], foreground=colors['fg'])
        self.style.configure('TLabelframe', background=colors['bg'], bordercolor=colors['border_color'])
        self.style.configure('TLabelframe.Label', background=colors['bg'], foreground=colors['primary'])
        
        # Entry and Combobox styles
        self.style.configure('TEntry', fieldbackground=colors['textfield'], foreground=colors['fg'])
        self.style.configure('TCombobox', fieldbackground=colors['textfield'], foreground=colors['fg'])
        
        # Scrollbar styles
        self.style.configure('TScrollbar', background=colors['primary'], troughcolor=colors['surface'])
        
        # Checkbutton and Radiobutton styles
        self.style.configure('TCheckbutton',
                             background=colors['bg'],
                             foreground=colors['fg'],
                             indicatorbackground=colors['surface'],
                             indicatorforeground=colors['primary'])
        self.style.map('TCheckbutton',
                       background=[('active', colors['bg'])],
                       foreground=[('active', colors['fg'])])

        self.style.configure('TRadiobutton',
                             background=colors['bg'],
                             foreground=colors['fg'],
                             indicatorbackground=colors['surface'],
                             indicatorforeground=colors['primary'])
        self.style.map('TRadiobutton',
                       background=[('active', colors['bg'])],
                       foreground=[('active', colors['fg'])])

        # OptionMenu (TMenubutton) style
        self.style.configure('TMenubutton',
                             background=colors['button_bg'],
                             foreground=colors['fg'],
                             bordercolor=colors['primary'],
                             lightcolor=colors['primary'],
                             darkcolor=colors['primary'],
                             arrowcolor=colors['fg'],
                             padding=5,
                             borderradius=5) # Rounded corners
        self.style.map('TMenubutton',
                       background=[('active', colors['button_hover_bg'])],
                       foreground=[('active', colors['fg'])])
        
        # Notebook (tabs) styles
        self.style.configure('TNotebook',
                             background=colors['bg'],
                             tabposition='nw',
                             padding=[5, 5, 5, 5])
        self.style.configure('TNotebook.Tab',
                             background=colors['surface'],
                             foreground=colors['fg'],
                             lightcolor=colors['surface'],
                             darkcolor=colors['surface'],
                             bordercolor=colors['primary'],
                             padding=[10, 5],
                             borderradius=[5, 5, 0, 0] # Rounded top corners for tabs
                             )
        self.style.map('TNotebook.Tab',
                       background=[('selected', colors['primary']), ('active', colors['button_hover_bg'])],
                       foreground=[('selected', colors['fg']), ('active', colors['fg'])])
        
        # Treeview (HistoryPage list) styles
        self.style.configure('Treeview.Heading',
                             background=colors['primary'],
                             foreground=colors['fg'],
                             font=('Inter', 10, 'bold'))
        self.style.map('Treeview.Heading',
                       background=[('active', colors['button_hover_bg'])])

        self.style.configure('Treeview',
                             background=colors['surface'],
                             foreground=colors['fg'],
                             fieldbackground=colors['surface'],
                             rowheight=25,
                             bordercolor=colors['primary'],
                             relief='flat')
        self.style.map('Treeview',
                       background=[('selected', colors['selected_item_bg'])],
                       foreground=[('selected', colors['selected_item_fg'])])

        # Custom page-specific styles
        self.style.configure('DiagnosisPage.TFrame', background=colors['diagnosis_page_bg'])
        self.style.configure('DiagnosisPageTitle.TLabel',
                             background=colors['diagnosis_page_bg'],
                             foreground=colors['diagnosis_title_fg'])
        
        self.style.configure('AboutPage.TFrame', background=colors['about_page_bg'])
        self.style.configure('AboutPageTitle.TLabel',
                             background=colors['about_page_bg'],
                             foreground=colors['about_page_title_fg'])
        self.style.configure('AboutPageSectionTitle.TLabel',
                             background=colors['about_page_bg'],
                             foreground=colors['about_page_title_fg'],
                             font=('Inter', 13, 'bold'))
        self.style.configure('AboutPageContent.TLabel',
                             background=colors['about_page_bg'],
                             foreground=colors['about_page_text_fg'])


        # Card style (for help sections, etc.)
        self.style.configure('Card.TFrame',
                             background=colors['surface'],
                             borderwidth=1,
                             relief="solid",
                             bordercolor=colors['border_color'], # Use general border color
                             lightcolor=colors['border_color'],
                             darkcolor=colors['border_color'],
                             borderradius=8 # Rounded corners for cards
                            )

        # Transparent styles (for canvas-based elements like HomePage)
        self.style.configure('Transparent.TFrame', background='')
        self.style.configure('Transparent.TLabel', background='')


    def _update_widget_colors(self, colors):
        """
        Recursively updates the background and foreground colors of all Tkinter
        and ttk widgets based on the current theme. This ensures visual consistency.
        """
        def update_child_widgets(widget):
            try:
                # Update standard tkinter widgets
                if isinstance(widget, (tk.Frame, tk.LabelFrame)):
                    widget.config(bg=colors['bg'])
                elif isinstance(widget, tk.Canvas):
                    # Special handling for canvases based on their purpose
                    if hasattr(widget, 'is_diagnosis_canvas') and widget.is_diagnosis_canvas:
                        widget.config(bg=colors['diagnosis_page_bg'])
                    elif hasattr(widget, 'is_help_canvas') and widget.is_help_canvas:
                        widget.config(bg=colors['bg'])
                    elif hasattr(widget, 'is_about_canvas') and widget.is_about_canvas:
                        widget.config(bg=colors['about_page_bg'])
                    else: # Default canvas background
                        widget.config(bg=colors['bg'])
                elif isinstance(widget, tk.Label):
                    # Apply specific colors based on custom attributes set in page classes
                    if hasattr(widget, 'is_diagnosis_title_label') and widget.is_diagnosis_title_label:
                        widget.config(bg=colors['diagnosis_page_bg'], fg=colors['diagnosis_title_fg'])
                    elif hasattr(widget, 'is_homepage_element') and widget.is_homepage_element:
                        # Homepage elements on canvas need special handling for hover effects
                        # Reconfigure activebackground and activeforeground for consistency
                        widget.config(
                            fg=colors['homepage_fg'],
                            bg=colors['bg'], # Ensure base background matches theme
                            activebackground=colors['homepage_button_hover_bg'],
                            activeforeground=colors['homepage_button_hover_fg']
                        )
                        # Re-bind hover events with new colors
                        widget.bind("<Enter>", lambda e, l=widget: l.config(fg=colors['homepage_button_hover_fg'], bg=colors['homepage_button_hover_bg']))
                        widget.bind("<Leave>", lambda e, l=widget: l.config(fg=colors['homepage_fg'], bg=colors['bg']))
                    elif hasattr(widget, 'is_image_canvas_zoom_label') and widget.is_image_canvas_zoom_label:
                        widget.config(bg=colors['image_canvas_bg'], fg=colors['fg'])
                    elif hasattr(widget, 'is_help_title') and widget.is_help_title:
                        widget.config(foreground=colors['help_page_title_fg'])
                    elif hasattr(widget, 'is_help_subtitle') and widget.is_help_subtitle:
                        widget.config(foreground=colors['help_page_text_fg']) # Changed to help_page_text_fg for consistency
                    elif hasattr(widget, 'is_help_section_title') and widget.is_help_section_title:
                        widget.config(foreground=colors['help_page_text_fg'])
                    elif hasattr(widget, 'is_help_content') and widget.is_help_content:
                        widget.config(foreground=colors['help_page_text_fg'])
                    elif hasattr(widget, 'is_toc_link') and widget.is_toc_link:
                        widget.config(fg=colors['fg'], activebackground=colors['primary'], activeforeground=colors['fg'])
                    else: # Default label background/foreground
                        widget.config(bg=colors['bg'], fg=colors['fg'])
                elif isinstance(widget, tk.Text):
                    widget.config(bg=colors['surface'], fg=colors['fg'],
                                  insertbackground=colors['fg'])
                elif isinstance(widget, tk.Listbox):
                    widget.config(bg=colors['surface'], fg=colors['fg'],
                                  selectbackground=colors['primary'])
                # No need to update ttk widgets directly here, as their styles are configured via self.style.configure
                # The style system handles their updates automatically when self.style is reconfigured.

            except tk.TclError as e:
                # Catch TclError if widget is destroyed during update
                logging.debug(f"Widget {widget} destroyed during theme update: {e}")
            except Exception as e:
                logging.error(f"Error updating widget {widget} color: {e}")
            
            # Recursively update children
            for child in widget.winfo_children():
                update_child_widgets(child)
        
        # Update specific canvas items that are not widgets (e.g., text items on HomePage)
        if "HomePage" in self.frames:
            homepage_frame = self.frames["HomePage"]
            if homepage_frame.canvas:
                homepage_frame.canvas.itemconfigure("homepage_title", fill=colors['homepage_fg'])
                homepage_frame.canvas.itemconfigure("homepage_subtitle", fill=colors['homepage_fg'])
                homepage_frame.canvas.itemconfigure("homepage_footer", fill=colors['homepage_fg'])
                
                # Update the toggle theme button text based on the new mode
                if hasattr(homepage_frame, 'toggle_theme_button_label') and homepage_frame.toggle_theme_button_label:
                    homepage_frame.toggle_theme_button_label.config(text="🌙" if self.dark_mode else "☀️")
                
                # Update overlay color if it exists
                if hasattr(homepage_frame, 'overlay_id') and homepage_frame.overlay_id:
                    # For a subtle overlay, we can adjust the stipple or recreate with a semi-transparent image
                    # For simplicity, let's just ensure it's black with a stipple
                    homepage_frame.canvas.itemconfigure(homepage_frame.overlay_id, fill="black", stipple="gray25")


        # Update text area tag colors in DiagnosisPage
        if "DiagnosisPage" in self.frames:
            diagnosis_frame = self.frames["DiagnosisPage"]
            # These tags use semantic colors which should generally remain consistent regardless of theme
            # However, if 'fg' is used, it should be updated.
            diagnosis_frame.result_text_area.tag_configure("header", foreground="#1B5E20" if not self.dark_mode else colors['fg']) # Keep dark green for light, use fg for dark
            diagnosis_frame.result_text_area.tag_configure("disease_icon", foreground=colors['fg'])
            # Confidence colors remain static (green, orange, red) as they are semantic
            diagnosis_frame.result_text_area.tag_configure("separator", foreground="#CCCCCC")
            diagnosis_frame.result_text_area.tag_configure("treatment_header", foreground="#6C757D" if not self.dark_mode else colors['fg']) # Keep grey for light, use fg for dark
            diagnosis_frame.result_text_area.tag_configure("disease_description_header", foreground="#007BFF" if not self.dark_mode else colors['fg']) # Keep blue for light, use fg for dark

        # Start the recursive update from the root window
        update_child_widgets(self)
        
        # Add a subtle visual transition effect
        self._animate_theme_transition()

    def _animate_theme_transition(self):
        """Adds a subtle fade-in transition effect when the theme changes."""
        # Temporarily set alpha to slightly transparent, then fade back to opaque
        self.attributes('-alpha', 0.9)
        self.after(50, lambda: self.attributes('-alpha', 1.0)) # Fade back in quickly

    def _save_theme_preference(self):
        """Saves the current theme preference to a configuration file."""
        config_path = os.path.join(os.path.expanduser('~'), '.crop_diagnosis_theme.cfg')
        try:
            with open(config_path, 'w') as f:
                f.write('dark' if self.dark_mode else 'light')
            logging.info(f"Theme preference saved: {'dark' if self.dark_mode else 'light'}")
        except Exception as e:
            logging.warning(f"Couldn't save theme preference to {config_path}: {e}")

    def _show_notification(self, message, message_type="info", duration=3000):
        """
        Displays a temporary, non-blocking notification at the top of the main window.
        message_type can be "info", "success", "warning", "error".
        """
        # Clear any existing notification timer
        if self.notification_timer:
            self.after_cancel(self.notification_timer)
            self.notification_timer = None

        # Destroy previous notification label if it exists
        if self.notification_label:
            self.notification_label.destroy()
            self.notification_label = None

        # Determine background and foreground colors based on message type
        bg_color = self.colors.get('surface', '#FFFFFF') # Default to white if colors not loaded
        fg_color = self.colors.get('fg', '#212121')     # Default to dark grey

        if message_type == "success":
            bg_color = "#4CAF50"  # Green
            fg_color = "white"
        elif message_type == "warning":
            bg_color = "#FFC107"  # Amber
            fg_color = "black"
        elif message_type == "error":
            bg_color = "#F44336"  # Red
            fg_color = "white"
        elif message_type == "info":
            bg_color = "#2196F3" # Blue
            fg_color = "white"

        self.notification_label = tk.Label(
            self,
            text=message,
            bg=bg_color,
            fg=fg_color,
            font=("Inter", 10, "bold"), # Use 'Inter' font
            relief="flat",
            padx=10,
            pady=5,
            wraplength=self.winfo_width() * 0.7 # Wrap text if window is narrow
        )
        # Place the label at the top of the window, centered horizontally
        self.notification_label.place(relx=0.5, rely=0, anchor="n", relwidth=0.8)
        
        # Schedule the notification to hide after 'duration' milliseconds
        self.notification_timer = self.after(duration, self._hide_notification)
        logging.info(f"Notification displayed: [{message_type.upper()}] {message}")

    def _hide_notification(self):
        """Hides and destroys the current notification."""
        if self.notification_label:
            self.notification_label.destroy()
            self.notification_label = None
        if self.notification_timer:
            self.after_cancel(self.notification_timer)
            self.notification_timer = None
        logging.debug("Notification hidden.")

    def _load_theme_preference(self):
        """Loads the saved theme preference from a configuration file."""
        config_path = os.path.join(os.path.expanduser('~'), '.crop_diagnosis_theme.cfg')
        try:
            if os.path.exists(config_path):
                with open(config_path, 'r') as f:
                    self.dark_mode = f.read().strip() == 'dark'
                logging.info(f"Theme preference loaded: {'dark' if self.dark_mode else 'light'}")
            else:
                # If no preference file exists, default to light mode
                self.dark_mode = False 
                logging.info("No theme preference file found. Defaulting to light mode.")
        except Exception as e:
            logging.warning(f"Couldn't load theme preference from {config_path}: {e}. Defaulting to light mode.")
            self.dark_mode = False 

    def _get_data_file_path(self):
        """Returns the path to the application data file."""
        app_data_dir = os.path.join(os.path.expanduser('~'), '.crop_diagnosis_app_data')
        os.makedirs(app_data_dir, exist_ok=True) # Ensure directory exists
        return os.path.join(app_data_dir, 'app_data.json')

    def _load_app_data(self):
        """Loads application data (history, theme) from a JSON file."""
        data_file = self._get_data_file_path()
        try:
            if os.path.exists(data_file):
                with open(data_file, 'r', encoding='utf-8') as f:
                    data = json.load(f)
                    self.dark_mode = data.get('dark_mode', False) # Load theme preference
                    
                    # Filter out non-existent image paths from history to avoid errors
                    loaded_history = data.get('diagnosis_history', [])
                    self.diagnosis_history = [
                        item for item in loaded_history 
                        if 'image' not in item or (item['image'] and os.path.exists(item['image']))
                    ]
                    # Filter recent images, ensuring paths exist
                    self.recent_images = [
                        path for path in data.get('recent_images', [])
                        if os.path.exists(path)
                    ]
                    logging.info("Application data loaded successfully.")
            else:
                logging.info("No application data file found. Starting with default settings.")
        except json.JSONDecodeError as e:
            logging.error(f"Error decoding application data JSON from {data_file}: {e}. Data might be corrupted.", exc_info=True)
            self._show_notification("Corrupted application data file. Starting with default settings.", "warning")
            # Reset data on corruption
            self.diagnosis_history = []
            self.recent_images = []
            self.dark_mode = False
        except Exception as e:
            logging.error(f"Error loading application data from {data_file}: {e}", exc_info=True)
            self._show_notification(f"Failed to load application data: {e}. Starting with default settings.", "error")
            # Reset data on other errors
            self.diagnosis_history = []
            self.recent_images = []
            self.dark_mode = False

    def _save_app_data(self):
        """Saves application data (history, theme) to a JSON file."""
        data_file = self._get_data_file_path()
        data = {
            'dark_mode': self.dark_mode,
            'diagnosis_history': self.diagnosis_history,
            'recent_images': self.recent_images
        }
        try:
            with open(data_file, 'w', encoding='utf-8') as f:
                json.dump(data, f, indent=4)
            logging.info(f"Application data saved successfully to {data_file}.")
        except Exception as e:
            logging.error(f"Error saving application data to {data_file}: {e}", exc_info=True)
            self._show_notification(f"Failed to save application data: {e}", "error")

    def show_frame(self, page_name):
        '''
        Raises the specified frame to the top, making it visible.
        Calls the 'on_show' method of the frame if it exists.
        '''
        frame = self.frames.get(page_name)
        if frame:
            frame.tkraise()
            if hasattr(frame, "on_show"):
                frame.on_show()
            logging.info(f"Switched to page: {page_name}")
        else:
            logging.error(f"Attempted to show non-existent frame: {page_name}")
            self._show_notification(f"Error: Page '{page_name}' not found.", "error")

    def quit(self):
        """
        Saves application data before gracefully quitting the application.
        """
        logging.info("Application quitting. Saving data...")
        self._save_app_data()
        self.destroy() # Destroy the main Tkinter window
        logging.info("Application closed.")

    # Improved non-crop detection methods
    def is_non_crop_image(self, img_path):
        """
        Detects if an image is likely not a crop leaf image using either:
        1. ML classifier (if available and trained for this purpose)
        2. Enhanced heuristic approach (fallback)
        """
        if self.crop_classifier:
            # If a dedicated crop classifier model is loaded, use it.
            # This assumes the model is trained to distinguish crop leaves from other objects.
            return self._ml_non_crop_detection(img_path)
        else:
            # Fallback to heuristic detection if no ML classifier is available.
            return self._heuristic_non_crop_detection(img_path)

    def _ml_non_crop_detection(self, img_path):
        """
        Uses a trained machine learning classifier to detect non-crop images.
        Assumes the 'crop_classifier' model is loaded and configured.
        """
        try:
            # Load and preprocess image for the crop classifier model
            # Assuming the crop classifier expects a target_size of (224, 224)
            img = image.load_img(image_path, target_size=(224, 224))
            img_array = image.img_to_array(img) / 255.0 # Normalize pixel values
            img_array = np.expand_dims(img_array, axis=0) # Add batch dimension
            
            # Predict (assuming model outputs a probability, e.g., 0 for non-crop, 1 for crop)
            prediction = self.crop_classifier.predict(img_array, verbose=0)
            # Return True if the prediction confidence for "non-crop" is high (e.g., > 0.5)
            # The threshold (0.5) can be adjusted based on model performance and desired sensitivity.
            is_non_crop = prediction[0][0] < 0.6 
            logging.info(f"ML non-crop detection for {img_path}: {is_non_crop} (Raw prediction: {prediction[0][0]:.2f})")
            return is_non_crop
            
        except Exception as e:
            logging.error(f"ML non-crop detection failed for {img_path}: {e}", exc_info=True)
            # Default to assuming it's a crop if ML detection fails, to allow diagnosis to proceed.
            return False

    def _heuristic_non_crop_detection(self, img_path):
        """
        Enhanced heuristic approach for non-crop detection based on image properties.
        This is a fallback if no ML crop classifier is available.
        """
        try:
            img = Image.open(img_path).convert("RGB") # Ensure RGB for consistent processing
            img_array = np.array(img)
            
            # Check minimum image dimensions early
            if img.size[0] < 100 or img.size[1] < 100: # Smaller threshold than 200x200 for a hard rejection
                logging.info(f"Heuristic non-crop detection for {img_path}: True (Reason: Image too small - {img.size[0]}x{img.size[1]})")
                return True

            # Convert to HSV for color analysis (Hue, Saturation, Value)
            hsv = cv2.cvtColor(img_array, cv2.COLOR_RGB2HSV)
            
            # 1. Green color dominance: Check for a significant amount of green pixels.
            # Green hue range (approx 30-90 in OpenCV HSV)
            # Increased saturation threshold to filter out dull greens
            green_pixels_mask = (hsv[:,:,0] > 30) & (hsv[:,:,0] < 90) & (hsv[:,:,1] > 40) 
            green_ratio = np.sum(green_pixels_mask) / (img.size[0] * img.size[1])
            
            # 2. Texture analysis using Laplacian variance (higher variance means more texture)
            gray = cv2.cvtColor(img_array, cv2.COLOR_RGB2GRAY)
            laplacian_var = cv2.Laplacian(gray, cv2.CV_64F).var()
            
            # 3. Brightness and contrast analysis
            avg_brightness = np.mean(gray)
            contrast = gray.std()
            
            # Decision thresholds (tuned empirically, may need adjustment)
            # If any of these conditions are met, it's likely not a typical crop leaf.
            if (green_ratio < 0.10 or      # Not enough vibrant green (increased threshold)
                laplacian_var < 10 or      # Too smooth (lower threshold for more strictness)
                avg_brightness < 10 or     # Too dark (adjusted threshold)
                avg_brightness > 245 or    # Too bright (adjusted threshold)
                contrast < 15):            # Low contrast (adjusted threshold)
                logging.info(f"Heuristic non-crop detection for {img_path}: True (Green Ratio: {green_ratio:.2f}, Laplacian Var: {laplacian_var:.2f}, Brightness: {avg_brightness:.2f}, Contrast: {contrast:.2f})")
                return True
                
            logging.info(f"Heuristic non-crop detection for {img_path}: False (Green Ratio: {green_ratio:.2f}, Laplacian Var: {laplacian_var:.2f}, Brightness: {avg_brightness:.2f}, Contrast: {contrast:.2f})")
            return False
            
        except Exception as e:
            logging.error(f"Heuristic non-crop detection failed for {img_path}: {e}", exc_info=True)
            # Default to assuming it's a crop if heuristic detection fails.
            return False

# Import page classes here to ensure they are defined when App tries to create instances.
# This structure helps avoid circular imports if pages also need to import App.
from tkinter import ttk, filedialog, messagebox, Toplevel
from PIL import Image, ImageTk, ImageEnhance, ImageFilter, ImageOps, ImageGrab
import time, threading, os, sys, logging, json, shutil, tempfile
from datetime import datetime
from fpdf import FPDF
import numpy as np
import cv2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model

# Dummy classes for pages to allow App class to be defined without full page implementations.
# In a real scenario, these would be imported from separate files.
# For this response, they are included after App, but conceptually they are separate modules.

class HomePage(ttk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent)
        self.controller = controller
        self._ui_built = False
        self.toggle_theme_button_label = None
        self.homepage_button_labels = []

        self.canvas = tk.Canvas(self, highlightthickness=0, bg='white')
        self.canvas.pack(fill="both", expand=True)

        self.original_bg = None
        self.bg_image = None
        self.bg_image_id = None # Initialize to None
        self.overlay_id = None # Initialize overlay ID

        bg_path = None
        try:
            # Corrected path joining: only pass the relative path to get_resource_path
            # Assuming the image is in assets/futuristic-technology-concept.jpg
            bg_path = self.controller.get_resource_path(os.path.join("assets", r"C:\Users\ayush\OneDrive\Desktop\project 1\CropDiseaseApp\assets\futuristic-technology-concept.jpg"))
            
            if not os.path.exists(bg_path):
                raise FileNotFoundError(f"Background image not found at: {bg_path}")
                
            self.original_bg = Image.open(bg_path)
            self.bg_image = ImageTk.PhotoImage(self.original_bg)
            
            self.bg_image_id = self.canvas.create_image(0, 0, anchor="nw", image=self.bg_image)
            self.canvas.tag_lower(self.bg_image_id)
            
            # Create a semi-transparent overlay
            self.overlay_id = self.canvas.create_rectangle(0, 0, 0, 0, fill="black", stipple="gray25", tags="overlay")
            self.canvas.tag_raise(self.overlay_id) # Ensure overlay is above the background image
            
        except Exception as e:
            logging.error(f"Failed to load background image from {bg_path}: {e}")
            self.original_bg = None
            self.canvas.config(bg=self.controller.colors['bg'])

        self.bind("<Configure>", self.resize_background)

    def resize_background(self, event):
        self.canvas.config(width=event.width, height=event.height)

        if self.original_bg and self.bg_image_id:
            try:
                resized_bg = self.original_bg.resize((event.width, event.height), Image.Resampling.LANCZOS)
                self.bg_image = ImageTk.PhotoImage(resized_bg)
                self.canvas.itemconfig(self.bg_image_id, image=self.bg_image)
                self.canvas.coords(self.bg_image_id, 0, 0)
            except Exception as e:
                logging.error(f"Error resizing background image: {e}")
        
        # Update overlay size
        if self.overlay_id:
            self.canvas.coords(self.overlay_id, 0, 0, event.width, event.height)
            self.canvas.itemconfigure(self.overlay_id, state="normal") # Ensure overlay is visible
            
        if not self._ui_built:
            self.build_content()
            self._ui_built = True

        canvas_width = self.canvas.winfo_width()
        canvas_height = self.canvas.winfo_height()

        content_center_x = canvas_width / 2
        content_start_y = canvas_height * 0.2

        # Adjust logo position relative to content start
        logo_y = content_start_y - 80 # Move logo higher
        self.canvas.coords("homepage_logo", content_center_x, logo_y)

        title_y = logo_y + 100
        self.canvas.coords("homepage_title", content_center_x, title_y)

        subtitle_y = title_y + 50
        self.canvas.coords("homepage_subtitle", content_center_x, subtitle_y)

        button_start_y = subtitle_y + 80
        button_height = 40
        button_spacing = 20

        for i, item_id in enumerate(self.canvas.find_withtag("homepage_button")):
            btn_y = button_start_y + i * (button_height + button_spacing)
            self.canvas.coords(item_id, content_center_x, btn_y)

        if self.toggle_theme_button_label:
            toggle_y = button_start_y + len(self.homepage_button_labels) * (button_height + button_spacing) + 20
            self.canvas.coords("toggle_theme_button", content_center_x, toggle_y)

        footer_y = toggle_y + 70 if self.toggle_theme_button_label else button_start_y + (len(self.homepage_button_labels) + 1) * (button_height + button_spacing) + 40
        self.canvas.coords("homepage_footer", content_center_x, footer_y)


    def build_content(self):
        logo_path = None
        try:
            logo_path = self.controller.get_resource_path(os.path.join("assets", r"C:\Users\ayush\OneDrive\Desktop\project 1\CropDiseaseApp\assets\icons8-chaff-48.png"))
            if os.path.exists(logo_path):
                logo_img = Image.open(logo_path).resize((150, 150), Image.Resampling.LANCZOS) # Smaller logo
                self.logo_photo = ImageTk.PhotoImage(logo_img)
                self.canvas.create_image(
                    0, 0,
                    image=self.logo_photo, 
                    anchor="center", 
                    tags="homepage_logo"
                )
        except Exception as e:
            logging.warning(f"Logo load failed from {logo_path}: {e}")

        self.canvas.create_text(
            0, 0,
            text="Welcome to Crop Disease Diagnosis", 
            font=("Inter", 38, "bold"), # Slightly larger font
            fill=self.controller.colors['homepage_fg'],
            tags="homepage_title"
        )
       
        self.canvas.create_text(
            0, 0,
            text="AI-powered detection of plant leaf diseases",
            font=("Inter", 20), # Slightly larger font
            fill=self.controller.colors['homepage_fg'],
            tags="homepage_subtitle"
        )
        
        buttons_data = [
            ("Diagnose Crop", "DiagnosisPage"),
            ("View History", "HistoryPage"),
            ("Provide Feedback", "FeedbackPage"),
            ("Help & Guide", "HelpPage"),
            ("About", "AboutPage"),
            ("Exit Application", None, lambda: self.controller.quit())
        ]

        for text, page_name, *command in buttons_data:
            btn_command = command[0] if command else (lambda p=page_name: self.controller.show_frame(p))
            
            btn_label = tk.Label(
                self.canvas,
                text=text,
                font=("Inter", 16, "bold"), # Larger font for buttons
                borderwidth=0,
                relief='flat',
                padx=25, # Increased padding
                pady=18, # Increased padding
                bg=self.controller.colors['bg'],
                fg=self.controller.colors['homepage_fg'],
                activebackground=self.controller.colors['homepage_button_hover_bg'], # Match hover for active
                activeforeground=self.controller.colors['homepage_button_hover_fg'], # Match hover for active
                cursor="hand2"
            )
            btn_label.is_homepage_button_label = True
            btn_label.is_homepage_element = True

            self.homepage_button_labels.append(btn_label)

            btn_label.bind("<Button-1>", lambda e, cmd=btn_command: self.animate_button(cmd))
            
            btn_label.bind("<Enter>", lambda e, l=btn_label: l.config(
                fg=self.controller.colors['homepage_button_hover_fg'],
                bg=self.controller.colors['homepage_button_hover_bg']
            ))
            btn_label.bind("<Leave>", lambda e, l=btn_label: l.config(
                fg=self.controller.colors['homepage_fg'],
                bg=self.controller.colors['bg']
            ))

            self.canvas.create_window(
                0, 0,
                window=btn_label, 
                anchor="center", 
                tags="homepage_button"
            )

        self.toggle_theme_button_label = tk.Label(
            self.canvas,
            text="☀️" if not self.controller.dark_mode else "🌙",
            font=("Segoe UI Emoji", 16), # Larger emoji font
            borderwidth=0,
            relief='flat',
            width=3,
            bg=self.controller.colors['bg'],
            fg=self.controller.colors['homepage_fg'],
            activebackground=self.controller.colors['homepage_button_hover_bg'], # Match hover for active
            activeforeground=self.controller.colors['homepage_button_hover_fg'], # Match hover for active
            cursor="hand2"
        )
        self.toggle_theme_button_label.is_homepage_button_label = True
        self.toggle_theme_button_label.is_homepage_element = True

        self.toggle_theme_button_label.bind("<Button-1>", lambda e: self.controller.toggle_theme_mode())
        self.toggle_theme_button_label.bind("<Enter>", lambda e, l=self.toggle_theme_button_label: l.config(
            fg=self.controller.colors['homepage_button_hover_fg'],
            bg=self.controller.colors['homepage_button_hover_bg']
        ))
        self.toggle_theme_button_label.bind("<Leave>", lambda e, l=self.toggle_theme_button_label: l.config(
            fg=self.controller.colors['homepage_fg'],
            bg=self.controller.colors['bg']
        ))
        
        self.canvas.create_window(
            0, 0,
            window=self.toggle_theme_button_label,
            anchor="center",
            tags="toggle_theme_button"
        )

        self.canvas.create_text(
            0, 0,
            text="Grow smarter   Stay healthier   ",
            font=("Inter", 16, "italic"), # Slightly larger font
            fill=self.controller.colors['homepage_fg'],
            tags="homepage_footer"
        )

    def animate_button(self, command):
        try:
            self.bell()
            # Add a subtle visual animation for the button itself if desired
            # For example, briefly change its size or color, then revert
            # For now, just a slight delay before executing the command
            self.after(100, command)
        except Exception as e:
            logging.warning(f"Button animation failed: {e}")

    def on_show(self):
        # Update the toggle theme button text when the page is shown
        if self.toggle_theme_button_label:
            self.toggle_theme_button_label.config(text="🌙" if self.controller.dark_mode else "☀️")
        # Ensure background is correctly set if image loading failed
        if not self.original_bg:
            self.canvas.config(bg=self.controller.colors['bg'])


class DiagnosisPage(ttk.Frame):
    def __init__(self, parent, controller):
        ttk.Frame.__init__(self, parent)
        self.controller = controller
        self.uploaded_image = None
        self.original_img = None
        self.current_image_path = None
        self.filtered_image = None
        self.zoom_level = 1.0
        self.pan_start_x = 0
        self.pan_start_y = 0
        self.current_display_img = None
        self.loading_frames = ["⣾", "⣽", "⣻", "⢿", "⡿", "⣟", "⣯", "⣷"]
        self.loading_idx = 0
        self.loading_active = False
        self.last_treatment_suggestions = []
        self.last_diagnosis_details = {}

        self.crop_rect_id = None
        self.start_x = None
        self.start_y = None
        self.cropping_active = False
        self.cropped_image = None

        self.camera = None
        self.camera_active = False
        self.camera_window = None
        self.camera_label = None
        self.camera_frame_id = None

        self.disease_info = {
            "Apple_Apple_scab": {
                "description": "Apple scab is a common fungal disease that affects apple and crabapple trees. It causes olive-green to brown spots on leaves, fruit, and twigs. Severe infections can lead to premature leaf drop and deformed fruit.",
                "symptoms": "Olive-green to brown spots on leaves, often with a velvety texture. Spots on fruit are dark, circular, and may become corky or cracked. Twig lesions can also occur."
            },
            "Apple_Black_rot": {
                "description": "Black rot is a fungal disease that affects apple trees, causing lesions on leaves, cankers on branches, and a characteristic black rot on fruit. It can lead to significant yield losses.",
                "symptoms": "Leaf spots are purplish with a brown center. Cankers on branches are sunken and discolored. Fruit rot begins as a brown spot that rapidly expands and turns black, often with concentric rings."
            },
            "Apple_Cedar_apple_rust": {
                "description": "Cedar-apple rust is a fungal disease that requires two hosts to complete its life cycle: a cedar (or juniper) tree and an apple (or crabapple) tree. It causes bright orange spots on apple leaves and can deform fruit.",
                "symptoms": "Bright orange-yellow spots on apple leaves, often with small black dots (spermagonia) in the center. Galls may form on cedar trees, producing gelatinous orange horns in wet weather."
            },
            "Apple_healthy": {
                "description": "The apple leaf appears healthy with no visible signs of disease. Maintaining good cultural practices is key to preventing future issues.",
                "symptoms": "Uniform green coloration, no spots, lesions, or deformities. Leaves are turgid and vibrant."
            },
            "Blueberry_healthy": {
                "description": "The blueberry plant appears healthy. Continue with good plant care practices to maintain its health.",
                "symptoms": "Vibrant green leaves, no discoloration, spots, or wilting. Healthy berry development."
            },
            "Cherry_Powdery_mildew": {
                "description": "Powdery mildew is a fungal disease that appears as white, powdery patches on the surface of leaves, stems, and sometimes fruit. Severe infections can stunt growth and reduce yield.",
                "symptoms": "White, powdery spots on leaves, stems, and fruit. Leaves may curl, distort, or turn yellow. Young leaves are often more susceptible."
            },
            "Cherry_healthy": {
                "description": "The cherry plant appears healthy. Continue with good plant care practices to maintain its health.",
                "symptoms": "Vibrant green leaves, no discoloration, spots, or wilting. Healthy fruit development."
            },
            "Corn_Cercospora_leaf_spot_Gray_leaf_spot": {
                "description": "Cercospora leaf spot (also known as Gray leaf spot) is a fungal disease of corn characterized by rectangular, gray-to-tan lesions on leaves. Severe infections can lead to significant yield loss.",
                "symptoms": "Long, narrow, rectangular lesions (1-2 inches long) that are gray to tan in color. Lesions are typically restricted by leaf veins. May appear water-soaked initially."
            },
            "Corn_Common_rust": {
                "description": "Common rust is a fungal disease of corn characterized by the formation of reddish-brown pustules on leaves. Severe infections can reduce photosynthetic area and impact yield.",
                "symptoms": "Small, cinnamon-brown to reddish-brown pustules on both upper and lower leaf surfaces. Pustules may rupture, releasing powdery spores."
            },
            "Corn_Northern_Leaf_Blight": {
                "description": "Northern Leaf Blight is a fungal disease of corn that causes long, elliptical, gray-green lesions on leaves. It can significantly reduce photosynthetic area and affect grain fill.",
                "symptoms": "Long, elliptical, gray-green to tan lesions on leaves, typically 1 to 6 inches long. Lesions may coalesce, blighting large areas of the leaf."
            },
            "Corn_healthy": {
                "description": "The corn plant appears healthy. Continue with good plant care practices to maintain its health.",
                "symptoms": "Vibrant green leaves, no discoloration, spots, or wilting. Healthy stalk and ear development."
            },
            "Grape_Black_rot": {
                "description": "Black rot is a destructive fungal disease of grapes that affects leaves, shoots, and fruit. It causes characteristic black, shriveled mummies on the fruit.",
                "symptoms": "Small, circular, reddish-brown spots on leaves that enlarge and turn tan with dark borders. Black, shriveled, raisin-like fruit mummies. Lesions on shoots and tendrils."
            },
            "Grape_Esca_Black_Measles": {
                "description": "Esca (also known as Black Measles) is a complex of fungal diseases affecting grapevines, leading to wood decay and foliar symptoms. It can cause sudden vine collapse or chronic decline.",
                "symptoms": "Foliar symptoms include interveinal chlorosis (yellowing) followed by necrosis (browning), often with a 'tiger-stripe' pattern. Fruit may develop dark spots and shrivel. Wood symptoms include dark streaking and decay."
            },
            "Grape_Leaf_blight": {
                "description": "Grape leaf blight is a general term for various conditions causing browning and death of leaf tissue. It can be caused by fungi, bacteria, or environmental factors.",
                "symptoms": "Irregular brown spots or patches on leaves, often starting at the margins or between veins. Affected areas may dry out and become brittle. Severe cases can lead to defoliation."
            },
            "Grape_healthy": {
                "description": "The grape plant appears healthy. Continue with good plant care practices to maintain its health.",
                "symptoms": "Vibrant green leaves, no discoloration, spots, or wilting. Healthy vine and fruit development."
            },
            "Orange_Haunglongbing_Citrus_greening": {
                "description": "Huanglongbing (HLB), also known as Citrus Greening, is a devastating bacterial disease of citrus trees spread by psyllids. It causes yellowing of leaves, misshapen fruit, and eventual tree decline.",
                "symptoms": "Asymmetrical blotchy mottling or yellowing of leaves, often resembling nutrient deficiencies but not symmetrical. Small, lopsided, green-bottomed fruit that taste bitter. Premature fruit drop."
            },
            "Peach_Bacterial_spot": {
                "description": "Bacterial spot is a common disease of peaches caused by Xanthomonas arboricola pv. pruni. It causes small, angular spots on leaves, lesions on twigs, and fruit spots.",
                "symptoms": "Small, angular, water-soaked spots on leaves that turn purplish-brown and may drop out, giving a 'shot-hole' appearance. Sunken, dark lesions on twigs and fruit. Fruit spots are dark, pitted, and may crack."
            },
            "Peach_healthy": {
                "description": "The peach plant appears healthy. Continue with good plant care practices to maintain its health.",
                "symptoms": "Vibrant green leaves, no discoloration, spots, or wilting. Healthy fruit development."
            },
            "Pepper_bell_Bacterial_spot": {
                "description": "Bacterial spot is a common and destructive disease of pepper plants caused by Xanthomonas bacteria. It leads to dark, water-soaked spots on leaves and fruit.",
                "symptoms": "Small, dark, water-soaked spots on leaves that may develop yellow halos. Spots on fruit are dark, raised, and scabby. Severe infections can cause leaf yellowing and defoliation."
            },
            "Pepper_bell_healthy": {
                "description": "The bell pepper plant appears healthy. Continue with good plant care practices to maintain its health.",
                "symptoms": "Vibrant green leaves, no discoloration, spots, or wilting. Healthy fruit development."
            },
            "Potato_Early_blight": {
                "description": "Early blight is a fungal disease affecting potato and tomato plants. It causes dark, concentric ringed spots on older leaves, leading to defoliation and reduced yield.",
                "symptoms": "Dark brown to black spots with concentric rings (like a target) on older leaves. Lesions may also appear on stems and tubers. Yellowing of tissue around spots."
            },
            "Potato_Late_blight": {
                "description": "Late blight is a devastating disease of potato and tomato caused by a water mold (Phytophthora infestans). It can rapidly destroy entire crops under cool, wet conditions.",
                "symptoms": "Irregular, water-soaked, dark green to brown lesions on leaves, often starting at the leaf tips or edges. White, fuzzy fungal growth may be visible on the undersides of leaves during humid conditions. Brown, firm rot on tubers."
            },
            "Potato_healthy": {
                "description": "The potato plant appears healthy. Continue with good plant care practices to maintain its health.",
                "symptoms": "Vibrant green leaves, no discoloration, spots, or wilting. Healthy tuber development."
            },
            "Raspberry_healthy": {
                "description": "The raspberry plant appears healthy. Continue with good plant care practices to maintain its health.",
                "symptoms": "Vibrant green leaves, no discoloration, spots, or wilting. Healthy berry development."
            },
            "Soybean_healthy": {
                "description": "The soybean plant appears healthy. Continue with good plant care practices to maintain its health.",
                "symptoms": "Vibrant green leaves, no discoloration, spots, or wilting. Healthy pod development."
            },
            "Squash_Powdery_mildew": {
                "description": "Powdery mildew is a common fungal disease affecting squash and other cucurbits. It appears as white, powdery spots on leaves and stems, reducing photosynthesis and yield.",
                "symptoms": "White, powdery patches on the upper and lower surfaces of leaves and stems. Leaves may turn yellow, then brown, and eventually die. Reduced fruit size and quality."
            },
            "Strawberry_Leaf_scorch": {
                "description": "Leaf scorch is a fungal disease of strawberries causing purplish spots on leaves that eventually enlarge and coalesce, leading to a 'scorched' appearance.",
                "symptoms": "Small, purplish spots on leaves that enlarge into irregular, reddish-brown blotches. Margins of the spots may turn brown or reddish. Severe infections can cause leaves to dry up and curl."
            },
            "Strawberry_healthy": {
                "description": "The strawberry plant appears healthy. Continue with good plant care practices to maintain its health.",
                "symptoms": "Vibrant green leaves, no discoloration, spots, or wilting. Healthy fruit development."
            },
            "Tomato_Bacterial_spot": {
                "description": "Bacterial spot is a common and destructive disease of tomato and pepper caused by several species of Xanthomonas bacteria. It leads to dark, water-soaked spots on leaves and fruit.",
                "symptoms": "Small, dark, water-soaked spots on leaves that may develop yellow halos. Spots on fruit are dark, raised, and scabby. Severe infections can cause leaf yellowing and defoliation."
            },
            "Tomato_Early_blight": {
                "description": "Early blight is a fungal disease affecting tomato and potato plants. It causes dark, concentric ringed spots on older leaves, leading to defoliation and reduced yield.",
                "symptoms": "Dark brown to black spots with concentric rings (like a target) on older leaves. Lesions may also appear on stems and fruit. Yellowing of tissue around spots."
            },
            "Tomato_Late_blight": {
                "description": "Late blight is a devastating disease of tomato and potato caused by a water mold (Phytophthora infestans). It can rapidly destroy entire crops under cool, wet conditions.",
                "symptoms": "Irregular, water-soaked, dark green to brown lesions on leaves, often starting at the leaf tips or edges. White, fuzzy fungal growth may be visible on the undersides of leaves during humid conditions. Brown, firm rot on fruit."
            },
            "Tomato_Leaf_Mold": {
                "description": "Leaf mold is a fungal disease of tomatoes, especially common in humid conditions. It causes velvety, olive-green to brown patches on the undersides of leaves.",
                "symptoms": "Yellowish spots on the upper leaf surface, with olive-green to brown, velvety fungal growth on the corresponding undersides. Leaves may curl, dry up, and fall off."
            },
            "Tomato_Septoria_leaf_spot": {
                "description": "Septoria leaf spot is a common fungal disease of tomatoes. It causes numerous small, circular spots on older leaves, often with dark borders and tiny black dots in the center.",
                "symptoms": "Numerous small, circular spots (1/8 to 1/4 inch) on older leaves. Spots have dark brown borders and tan to gray centers, often with tiny black specks (pycnidia) in the middle. Severe infections lead to defoliation."
            },
            "Tomato_Spider_mites _Two_spotted_spider_mite": {
                "description": "Spider mites are tiny pests that feed on plant cells, causing stippling and yellowing of leaves. Heavy infestations can lead to webbing and severe plant damage.",
                "symptoms": "Tiny yellow or white stippling (pinprick dots) on leaves. Yellowing, bronzing, or drying of leaves. Fine webbing on the undersides of leaves or between stems. Mites are barely visible to the naked eye."
            },
            "Tomato_Target_Spot": {
                "description": "Target spot is a fungal disease of tomatoes causing circular lesions with concentric rings, resembling a target. It affects leaves, stems, and fruit.",
                "symptoms": "Small, circular, water-soaked spots on leaves that enlarge and develop concentric rings, turning brown to black. A yellow halo may surround the spots. Lesions can also appear on stems and fruit."
            },
            "Tomato_Tomato_Yellow_Leaf_Curl_Virus": {
                "description": "Tomato Yellow Leaf Curl Virus (TYLCV) is a devastating viral disease of tomatoes transmitted by whiteflies. It causes severe stunting and yellowing of leaves.",
                "symptoms": "Severe stunting of plants. Upward curling and yellowing of leaf margins, especially on younger leaves. Leaves become thickened and brittle. Flowers may drop, and fruit set is severely reduced."
            },
            "Tomato_Tomato_mosaic_virus": {
                "description": "Tomato Mosaic Virus (ToMV) is a highly contagious viral disease that causes mosaic patterns, mottling, and distortion of leaves in tomato plants.",
                "symptoms": "Light and dark green mosaic patterns or mottling on leaves. Leaves may be distorted, puckered, or fern-like. Stunting of plants and reduced fruit size or quality."
            },
            "Tomato_healthy": {
                "description": "The tomato leaf appears healthy with no visible signs of disease. Continue with good plant care practices.",
                "symptoms": "Vibrant green leaves, no discoloration, spots, or wilting. Healthy stem and fruit development."
            }
        }


        self.grid_rowconfigure(0, weight=1)
        self.grid_columnconfigure(0, weight=1)

        self.canvas = tk.Canvas(self, borderwidth=0, highlightthickness=0)
        self.canvas.is_diagnosis_canvas = True
        self.scrollbar = ttk.Scrollbar(self, orient="vertical", command=self.canvas.yview)
        self.canvas.configure(yscrollcommand=self.scrollbar.set)
        
        self.scrollable_frame = ttk.Frame(self.canvas)
        self.canvas_frame = self.canvas.create_window((0,0), window=self.scrollable_frame, anchor="nw")

        self.scrollable_frame.bind("<Configure>", lambda e: self.canvas.configure(scrollregion=self.canvas.bbox("all")))
        self.bind("<Configure>", self._resize_canvas)

        self.canvas.grid(row=0, column=0, sticky="nsew")
        self.scrollbar.grid(row=0, column=1, sticky="ns")
        self.grid_rowconfigure(0, weight=1)
        self.grid_columnconfigure(0, weight=1)
        
        # Configure scrollable_frame columns for main content layout
        self.scrollable_frame.grid_columnconfigure(0, weight=1)
        self.scrollable_frame.grid_columnconfigure(1, weight=1)

        self.canvas.bind("<MouseWheel>", self._on_mousewheel)
        self.scrollable_frame.bind("<MouseWheel>", self._on_mousewheel)
        self.canvas.bind("<Button-4>", self._on_mousewheel)
        self.canvas.bind("<Button-5>", self._on_mousewheel)
        self.scrollable_frame.bind("<Button-4>", self._on_mousewheel)
        self.scrollable_frame.bind("<Button-5>", self._on_mousewheel)

        self.init_layout(self.scrollable_frame)
        
        # Ensure scrollregion is updated after all widgets are placed
        self.scrollable_frame.update_idletasks()
        self.canvas.configure(scrollregion=self.canvas.bbox("all"))

    def _resize_canvas(self, event):
        canvas_width = event.width
        # Update the width of the frame inside the canvas to match canvas width
        self.canvas.itemconfig(self.canvas_frame, width=canvas_width)
        # Reconfigure scrollregion after frame size might have changed
        self.scrollable_frame.update_idletasks()
        self.canvas.config(scrollregion=(0, 0, self.scrollable_frame.winfo_width(), self.scrollable_frame.winfo_reqheight()))
        
    def _on_mousewheel(self, event):
        if sys.platform == "darwin":
            self.canvas.yview_scroll(int(-1*(event.delta)), "units")
        elif sys.platform == "win32":
            self.canvas.yview_scroll(int(-1*(event.delta/120)), "units")
        else:
            if event.num == 4:
                self.canvas.yview_scroll(-1, "units")
            elif event.num == 5:
                self.canvas.yview_scroll(1, "units")


    def init_layout(self, parent):
        # Apply theme-specific background to the scrollable frame
        self.scrollable_frame.config(style='DiagnosisPage.TFrame')
        self.canvas.config(bg=self.controller.colors['diagnosis_page_bg'])

        # Page Title
        title_frame = ttk.Frame(parent, style='DiagnosisPage.TFrame')
        title_frame.grid(row=0, column=0, columnspan=2, pady=(0, 20), padx=10, sticky="ew")
        
        self.diagnosis_title_label = ttk.Label(
            title_frame, 
            text="Crop Disease Diagnosis", 
            font=("Inter", 18, "bold"),
            style='DiagnosisPageTitle.TLabel'
        )
        self.diagnosis_title_label.pack(side=tk.LEFT)

        self.diagnosis_title_label.bind("<Enter>", self._on_title_enter)
        self.diagnosis_title_label.bind("<Leave>", self._on_title_leave)
        
        # Image Tools Section (Upload, Camera, Clear, Screenshot, Crop)
        image_tools_frame = ttk.LabelFrame(parent, text="Image Tools", style='Section.TLabelframe')
        image_tools_frame.grid(row=1, column=0, columnspan=2, pady=(0, 10), padx=10, sticky="ew")
        
        # Control buttons within the image tools frame
        upload_button = ttk.Button(
            image_tools_frame,
            text="Upload Image",
            command=self.upload_image,
            style='TButton'
        )
        upload_button.pack(side=tk.LEFT, padx=5, pady=5)

        camera_button = ttk.Button(
            image_tools_frame,
            text="Live Camera",
            command=self.open_camera_feed,
            style='TButton'
        )
        camera_button.pack(side=tk.LEFT, padx=5, pady=5)

        clear_button = ttk.Button(
            image_tools_frame,
            text="Clear Image",
            command=self.clear_image,
            style='TButton'
        )
        clear_button.pack(side=tk.LEFT, padx=5, pady=5)
        
        self.preview_var = tk.BooleanVar(value=True)
        ttk.Checkbutton(
            image_tools_frame,
            text="Live Preview",
            variable=self.preview_var,
            command=self.toggle_live_preview,
            style='TCheckbutton'
        ).pack(side=tk.LEFT, padx=10, pady=5)

        screenshot_button = ttk.Button(
            image_tools_frame,
            text="Take Screenshot",
            command=self.take_screenshot,
            style='TButton'
        )
        screenshot_button.pack(side=tk.LEFT, padx=5, pady=5)

        self.crop_button = ttk.Button(
            image_tools_frame,
            text="Crop Image",
            command=self.activate_crop_mode,
            style='TButton',
            state=tk.DISABLED
        )
        self.crop_button.pack(side=tk.LEFT, padx=5, pady=5)

        self.apply_crop_button = ttk.Button(
            image_tools_frame,
            text="Apply Crop",
            command=self.apply_crop,
            style='TButton',
            state=tk.DISABLED
        )
        # Initially hide apply_crop_button
        self.apply_crop_button.pack_forget()

        self.reset_crop_button = ttk.Button(
            image_tools_frame,
            text="Reset Crop",
            command=self.reset_crop,
            style='TButton',
            state=tk.DISABLED
        )
        # Initially hide reset_crop_button
        self.reset_crop_button.pack_forget()

        # Status and Loading Indicator
        self.status_frame = ttk.Frame(parent, style='DiagnosisPage.TFrame')
        self.status_frame.grid(row=2, column=0, columnspan=2, pady=(0, 10), padx=10, sticky="ew")
        
        self.status_label = ttk.Label(
            self.status_frame,
            text="Ready to upload image",
            font=("Inter", 10),
            style='TLabel'
        )
        self.status_label.pack(side=tk.LEFT)
        
        self.loading_label = ttk.Label(
            self.status_frame,
            text="",
            font=("Inter", 10),
            style='TLabel'
        )
        self.loading_label.pack(side=tk.LEFT, padx=5)
        
        # Main Content Area (Image Preview and Diagnosis Results)
        # This frame will hold the image preview on the left and results on the right
        main_content_area = ttk.Frame(parent, style='DiagnosisPage.TFrame')
        main_content_area.grid(row=3, column=0, columnspan=2, sticky="nsew", padx=10, pady=10)
        main_content_area.grid_columnconfigure(0, weight=1, minsize=400) # Image Preview column
        main_content_area.grid_columnconfigure(1, weight=1, minsize=400) # Diagnosis Results column
        main_content_area.grid_rowconfigure(0, weight=1)


        # Image Preview Section
        self.image_frame = ttk.LabelFrame(main_content_area, text="Image Preview", style='Section.TLabelframe')
        self.image_frame.grid(row=0, column=0, sticky="nsew", padx=(0, 5), pady=5)
        self.image_frame.grid_rowconfigure(0, weight=1)
        self.image_frame.grid_columnconfigure(0, weight=1)
        
        self.image_canvas = tk.Canvas(self.image_frame, bg=self.controller.colors['image_canvas_bg'], highlightthickness=0)
        self.image_canvas.grid(row=0, column=0, sticky="nsew")
        
        self.image_hscroll = ttk.Scrollbar(self.image_frame, orient="horizontal", command=self.image_canvas.xview)
        self.image_vscroll = ttk.Scrollbar(self.image_frame, orient="vertical", command=self.image_canvas.yview)
        self.image_canvas.configure(xscrollcommand=self.image_hscroll.set, yscrollcommand=self.image_vscroll.set)
        
        self.image_hscroll.grid(row=1, column=0, sticky="ew")
        self.image_vscroll.grid(row=0, column=1, sticky="ns")
        
        self.zoom_label = ttk.Label(
            self.image_frame, 
            text="Zoom: 1.0x",
            background=self.controller.colors['zoom_label_bg'],
            foreground=self.controller.colors['fg']
        )
        self.zoom_label.is_image_canvas_zoom_label = True
        self.zoom_label.grid(row=1, column=0, sticky="se", padx=5, pady=5)

        self.compare_button = ttk.Button(
            self.image_frame,
            text="Compare Original",
            style='TButton'
        )
        self.compare_button.grid(row=1, column=0, sticky="sw", padx=5, pady=5)
        self.compare_button.bind("<ButtonPress-1>", self.start_compare_original)
        self.compare_button.bind("<ButtonRelease-1>", self.stop_compare_original)
        self.compare_button.config(state=tk.DISABLED)
        
        self.image_canvas.bind("<MouseWheel>", self.zoom_image)
        self.image_canvas.bind("<ButtonPress-1>", self.start_pan)
        self.image_canvas.bind("<B1-Motion>", self.pan_image)
        self.image_canvas.bind("<Configure>", self.update_scrollregion)
        
        # Diagnosis Results Section
        self.result_frame = ttk.LabelFrame(main_content_area, text="Diagnosis Results", style='Section.TLabelframe')
        self.result_frame.grid(row=0, column=1, sticky="nsew", padx=(5, 0), pady=5)
        self.result_frame.grid_rowconfigure(0, weight=1)
        self.result_frame.grid_columnconfigure(0, weight=1)

        self.result_text_area = tk.Text(
            self.result_frame,
            wrap=tk.WORD,
            font=("Inter", 12),
            padx=10,
            pady=10,
            height=10, # Initial height, will expand with grid weight
            state=tk.DISABLED
        )
        self.result_text_area.grid(row=0, column=0, sticky="nsew")
        result_scrollbar = ttk.Scrollbar(self.result_frame, command=self.result_text_area.yview)
        result_scrollbar.grid(row=0, column=1, sticky="ns")
        self.result_text_area.config(yscrollcommand=result_scrollbar.set)
        
        self.result_text_area.insert(tk.END, "No diagnosis available. Upload an image to start.")
        # These tags are configured in _update_widget_colors now for theme consistency
        self.result_text_area.tag_configure("header", font=("Inter", 14, "bold"))
        self.result_text_area.tag_configure("disease_icon", font=("Arial", 14))
        self.result_text_area.tag_configure("confidence_high", foreground="green")
        self.result_text_area.tag_configure("confidence_medium", foreground="orange")
        self.result_text_area.tag_configure("confidence_low", foreground="red")
        self.result_text_area.tag_configure("separator", foreground="#CCCCCC")
        self.result_text_area.tag_configure("treatment_header", font=("Inter", 11, "bold"))
        self.result_text_area.tag_configure("treatment_bullet", lmargin1=20, lmargin2=20)
        self.result_text_area.tag_configure("disease_description_header", font=("Inter", 11, "bold"))
        self.result_text_area.tag_configure("disease_description_content", lmargin1=20, lmargin2=20)

        # Progress Bar (positioned below main content area)
        self.progress = ttk.Progressbar(
            parent,
            orient="horizontal",
            length=300,
            mode="determinate"
        )
        self.progress.grid(row=4, column=0, columnspan=2, pady=(5, 10), padx=10, sticky="ew")
        self.progress.grid_remove() # Hide initially
        
        # Image Enhancement Tools Section
        self.create_filter_controls(parent)
        
        # Bottom Buttons (Save, Print, Back)
        button_frame = ttk.Frame(parent, style='DiagnosisPage.TFrame')
        button_frame.grid(row=6, column=0, columnspan=2, pady=10, padx=10, sticky="ew")
        
        self.save_button = ttk.Button(
            button_frame,
            text="Save Report",
            state=tk.DISABLED,
            command=self.save_report,
            style='TButton'
        )
        self.save_button.pack(side=tk.LEFT, padx=5, expand=True, fill=tk.X)

        self.print_current_report_button = ttk.Button(
            button_frame,
            text="Print Current Report",
            state=tk.DISABLED,
            command=self.print_current_report,
            style='TButton'
        )
        self.print_current_report_button.pack(side=tk.LEFT, padx=5, expand=True, fill=tk.X)
        
        back_button = ttk.Button(
            button_frame,
            text="Back to Home",
            command=lambda: self.controller.show_frame("HomePage"),
            style='TButton'
        )
        back_button.pack(side=tk.LEFT, padx=5, expand=True, fill=tk.X)
        
        # Configure parent grid columns to expand
        parent.grid_columnconfigure(0, weight=1)
        parent.grid_columnconfigure(1, weight=1)

    def _on_title_enter(self, event):
        self.diagnosis_title_label.config(foreground=self.controller.colors['diagnosis_title_hover_fg'])

    def _on_title_leave(self, event):
        self.diagnosis_title_label.config(foreground=self.controller.colors['diagnosis_title_fg'])


    def open_camera_feed(self):
        if self.camera_active:
            messagebox.showinfo("Camera Active", "Camera feed is already open.")
            return

        self.camera_window = tk.Toplevel(self.controller)
        self.camera_window.title("Live Camera Feed")
        self.camera_window.geometry("640x520")
        self.camera_window.protocol("WM_DELETE_WINDOW", self.close_camera_feed)

        camera_frame_controls = ttk.Frame(self.camera_window)
        camera_frame_controls.pack(pady=10)

        self.camera_label = ttk.Label(camera_frame_controls)
        self.camera_label.pack()

        capture_button = ttk.Button(
            camera_frame_controls,
            text="Capture Image",
            command=self.capture_image_from_camera,
            style='TButton'
        )
        capture_button.pack(pady=5)

        close_button = ttk.Button(
            camera_frame_controls,
            text="Close Camera",
            command=self.close_camera_feed,
            style='TButton'
        )
        close_button.pack(pady=5)

        try:
            self.camera = cv2.VideoCapture(0)
            if not self.camera.isOpened():
                raise IOError("Cannot open webcam")
            self.camera_active = True
            self.update_camera_feed()
        except Exception as e:
            messagebox.showerror("Camera Error", f"Could not open camera: {e}")
            logging.error(f"Camera initialization error: {e}")
            self.close_camera_feed()

    def update_camera_feed(self):
        if self.camera_active and self.camera and self.camera.isOpened():
            ret, frame = self.camera.read()
            if ret:
                cv2_image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                pil_image = Image.fromarray(cv2_image)
                
                max_width = 600
                max_height = 400
                pil_image.thumbnail((max_width, max_height), Image.Resampling.LANCZOS)

                self.camera_photo = ImageTk.PhotoImage(image=pil_image)
                self.camera_label.config(image=self.camera_photo)
                self.camera_label.image = self.camera_photo

                self.camera_window.after(10, self.update_camera_feed)
            else:
                logging.warning("Failed to read frame from camera.")
                self.close_camera_feed()
        else:
            self.close_camera_feed()


    def capture_image_from_camera(self):
        if self.camera_active and self.camera and self.camera.isOpened():
            ret, frame = self.camera.read()
            if ret:
                cv2_image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                pil_image = Image.fromarray(cv2_image)

                temp_file_path = os.path.join(tempfile.gettempdir(), f"captured_image_{int(time.time())}.png")
                pil_image.save(temp_file_path)

                self.close_camera_feed()
                self.upload_image(temp_file_path)
            else:
                self.controller._show_notification("Failed to capture image from camera.", "error")
                logging.error("Failed to capture image from camera.")
        else:
            self.controller._show_notification("No live camera feed is active.", "warning")

    def close_camera_feed(self):
        if self.camera:
            self.camera.release()
            self.camera = None
        self.camera_active = False
        if self.camera_window:
            self.camera_window.destroy()
            self.camera_window = None
        self.camera_label = None
        self.camera_frame_id = None


    def upload_image(self, file_path=None):
        if file_path is None:
            file_types = [
                ("Image files", "*.jpg;*.jpeg;*.png;*.bmp"),
                ("All files", "*.*")
            ]
            
            file_path = filedialog.askopenfilename(
                title="Select Crop Image",
                filetypes=file_types
            )
        
        if file_path:
            try:
                img = Image.open(file_path)
                img.verify()
                
                img = Image.open(file_path)
                img = ImageOps.exif_transpose(img)
                
                if img.size[0] < 200 or img.size[1] < 200:
                    self.controller._show_notification(
                        "Image resolution is too low for accurate diagnosis. Minimum 200x200 pixels.", 
                        "warning",
                        duration=5000
                    )
                
                self.current_image_path = file_path
                self.uploaded_image = img.copy()
                self.original_img = img.copy()
                self.filtered_image = self.uploaded_image.copy()
                self.cropped_image = None
                
                self.display_image(self.uploaded_image)
                
                self.compare_button.config(state=tk.NORMAL)
                self.crop_button.config(state=tk.NORMAL)
                self.print_current_report_button.config(state=tk.DISABLED)
                self.save_button.config(state=tk.DISABLED)

                self.status_label.config(text="Image uploaded. Applying filters or click 'Diagnose'.")
                self.clear_results()

                self.start_processing(self.current_image_path)

                self.reset_filters_ui()

            except Exception as e:
                self.status_label.config(text=f"Error: {str(e)}")
                logging.error(f"Image upload error: {e}")
                self.controller._show_notification(
                    f"Failed to process image: {str(e)}. Please select a valid image file.", 
                    "error",
                    duration=5000
                )

    def clear_image(self):
        self.uploaded_image = None
        self.original_img = None
        self.current_image_path = None
        self.filtered_image = None
        self.cropped_image = None
        self.image_canvas.delete("all")
        self.zoom_level = 1.0
        self.zoom_label.config(text="Zoom: 1.0x")
        self.compare_button.config(state=tk.DISABLED)
        self.crop_button.config(state=tk.DISABLED)
        # Ensure crop buttons are hidden and reset
        self.apply_crop_button.pack_forget()
        self.reset_crop_button.pack_forget()
        self.cropping_active = False
        # Rebind pan events
        self.image_canvas.unbind("<ButtonPress-1>")
        self.image_canvas.unbind("<B1-Motion>")
        self.image_canvas.unbind("<ButtonRelease-1>")
        self.image_canvas.bind("<ButtonPress-1>", self.start_pan)
        self.image_canvas.bind("<B1-Motion>", self.pan_image)

        self.clear_results()
        self.result_text_area.config(state=tk.NORMAL)
        self.result_text_area.delete(1.0, tk.END)
        self.result_text_area.insert(tk.END, "No diagnosis available. Upload an image to start.")
        self.result_text_area.config(state=tk.DISABLED)
        self.save_button.config(state=tk.DISABLED)
        self.status_label.config(text="Ready to upload image")
        self.reset_filters_ui()
        self.controller._show_notification("Image cleared.", "info")


    def take_screenshot(self):
        try:
            self.controller.withdraw()
            self.update_idletasks()

            x = self.controller.winfo_x()
            y = self.controller.winfo_y()
            width = self.controller.winfo_width()
            height = self.controller.winfo_height()

            bbox = (x, y, x + width, y + height)

            screenshot = ImageGrab.grab(bbox=bbox)

            self.controller.deiconify()

            default_filename = f"screenshot_{datetime.now().strftime('%Y%m%d_%H%M%S')}.png"
            file_path = filedialog.asksaveasfilename(
                defaultextension=".png",
                filetypes=[("PNG files", "*.png"), ("All files", "*.*")],
                initialfile=default_filename,
                title="Save Screenshot"
            )

            if file_path:
                screenshot.save(file_path)
                self.controller._show_notification(f"Screenshot saved to: {os.path.basename(file_path)}", "success")
            else:
                self.controller._show_notification("Screenshot saving cancelled.", "info")

        except Exception as e:
            self.controller.deiconify()
            self.controller._show_notification(f"Failed to take screenshot: {e}", "error")
            logging.error(f"Screenshot error: {e}")


    def start_compare_original(self, event):
        if self.original_img:
            self.display_image(self.original_img, is_original=True)

    def stop_compare_original(self, event):
        if self.filtered_image:
            self.display_image(self.filtered_image)

    def start_processing(self, file_path):
        self.clear_results()
        self.save_button.config(state=tk.DISABLED)
        self.print_current_report_button.config(state=tk.DISABLED)
        self.progress.grid()
        self.progress.start()
        self.status_label.config(text="Analyzing image...")
        
        self.loading_active = True
        self.animate_loading()
        
        threading.Thread(
            target=self.process_image,
            args=(file_path,),
            daemon=True
        ).start()

    def animate_loading(self):
        if self.loading_active:
            self.loading_label.config(text=self.loading_frames[self.loading_idx])
            self.loading_idx = (self.loading_idx + 1) % len(self.loading_frames)
            self.after(100, self.animate_loading)

    def process_image(self, image_path):
        temp_img_path = None
        try:
            if image_path not in self.controller.recent_images:
                self.controller.recent_images.insert(0, image_path)
                if len(self.controller.recent_images) > 20:
                    self.controller.recent_images.pop()
            
            start_time = time.time()
            img_to_diagnose = self.filtered_image
            
            if img_to_diagnose:
                temp_img_path = os.path.join(tempfile.gettempdir(), "temp_diagnosis_img.png")
                img_to_diagnose.save(temp_img_path)
                diagnosis_result = self.diagnose_disease(temp_img_path)
            else:
                diagnosis_result = "Error: No image to diagnose."

            processing_time = time.time() - start_time
            
            self.after(0, self.update_results, diagnosis_result, processing_time, image_path)
            
        except Exception as e:
            self.after(0, lambda: self.status_label.config(text=f"Error: {str(e)}"))
            logging.error(f"Processing error: {e}")
        finally:
            self.after(0, self.progress.stop)
            self.after(0, self.progress.grid_remove)
            self.loading_active = False
            self.after(0, lambda: self.loading_label.config(text=""))
            if temp_img_path and os.path.exists(temp_img_path):
                os.remove(temp_img_path)

    def update_results(self, prediction, processing_time, image_path):
        self.result_text_area.config(state=tk.NORMAL)
        self.result_text_area.delete(1.0, tk.END)
        
        if prediction.startswith("Error:"):
            self.result_text_area.insert(tk.END, "DIAGNOSIS FAILED\n\n", "header")
            self.result_text_area.insert(tk.END, prediction + "\n", "confidence_low")
            self.result_text_area.config(state=tk.DISABLED)
            self.status_label.config(text=f"Analysis failed: {prediction}")
            self.save_button.config(state=tk.DISABLED)
            self.print_current_report_button.config(state=tk.DISABLED)
            
            self.last_diagnosis_details = {}
            self.last_treatment_suggestions = []
            self.controller._show_notification("Diagnosis failed. See results for details.", "error")
            return
        
        result_lines = prediction.split("\n")
        
        self.result_text_area.insert(tk.END, "DIAGNOSIS RESULTS\n\n", "header")
        
        crop_type = "N/A"
        disease_name = "N/A"
        confidence = "N/A"

        for line in result_lines:
            if "Crop Type:" in line:
                crop_type = line.split(":")[1].strip()
                self.result_text_area.insert(tk.END, f"Crop Type: {crop_type}\n")
            elif "Disease:" in line:
                disease_name = line.split(":")[1].strip()
                icon = "⚠️" if "healthy" not in disease_name.lower() else "✅"
                self.result_text_area.insert(tk.END, f"{icon} Disease: {disease_name}\n", "disease_icon")
            elif "Confidence:" in line:
                confidence = line.split(":")[1].strip()
                try:
                    conf_val = float(confidence.replace("%", ""))
                    if conf_val > 75:
                        self.result_text_area.insert(tk.END, f"Confidence: {confidence}\n", "confidence_high")
                    elif conf_val > 50:
                        self.result_text_area.insert(tk.END, f"Confidence: {confidence}\n", "confidence_medium")
                    else:
                        self.result_text_area.insert(tk.END, f"Confidence: {confidence}\n")
                except ValueError:
                    self.result_text_area.insert(tk.END, f"Confidence: {confidence}\n")
            else:
                self.result_text_area.insert(tk.END, f"{line}\n")
        
        self.last_diagnosis_details = {
            "crop_type": crop_type,
            "disease_name": disease_name,
            "confidence": confidence
        }

        self.result_text_area.insert(tk.END, "\n" + "="*40 + "\n\n", "separator")
        self.result_text_area.insert(tk.END, "Disease Information:\n", "disease_description_header")
        
        full_disease_key = f"{crop_type}_{disease_name.replace(' ', '_')}"
        info = self.disease_info.get(full_disease_key, self.disease_info.get("healthy"))
        
        if info:
            self.result_text_area.insert(tk.END, f"Description: {info['description']}\n\n", "disease_description_content")
            self.result_text_area.insert(tk.END, f"Symptoms: {info['symptoms']}\n", "disease_description_content")
        else:
            self.result_text_area.insert(tk.END, "No detailed information available for this disease.\n", "disease_description_content")


        if "Disease:" in prediction:
            self.last_treatment_suggestions = self.add_treatment_suggestions(prediction)
            self.result_text_area.insert(tk.END, "\n" + "="*40 + "\n\n", "separator")
            self.result_text_area.insert(tk.END, "Treatment Suggestions:\n", "treatment_header")
            for suggestion in self.last_treatment_suggestions:
                self.result_text_area.insert(tk.END, f"- {suggestion}\n", "treatment_bullet")
        else:
            self.last_treatment_suggestions = []
        
        self.result_text_area.config(state=tk.DISABLED)

        self.status_label.config(
            text=f"Analysis complete in {processing_time:.2f}s | "
                 f"Image: {self.uploaded_image.size[0]}x{self.uploaded_image.size[1]}"
        )
        
        self.save_button.config(state=tk.NORMAL)
        self.print_current_report_button.config(state=tk.NORMAL)
        
        self.controller.diagnosis_history.append({
            "image": image_path,
            "diagnosis": prediction,
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "details": self.last_diagnosis_details,
            "treatment_suggestions": self.last_treatment_suggestions
        })
        
        if "HistoryPage" in self.controller.frames:
            self.controller.frames["HistoryPage"].refresh_history()
        
        self.controller._show_notification("Diagnosis complete!", "success")

    
    def add_treatment_suggestions(self, prediction):
        treatments = {
            "Apple_scab": [
                "Apply fungicides containing myclobutanil or sulfur",
                "Remove and destroy infected leaves and fruit",
                "Improve air circulation through pruning"
            ],
            "Black_rot": [
                "Prune out infected plant parts (canes, fruit)",
                "Apply fungicides like captan or mancozeb during susceptible stages",
                "Ensure good air circulation and sunlight exposure"
            ],
            "Cedar_apple_rust": [
                "Remove nearby cedar trees (alternate host)",
                "Apply fungicides such as myclobutanil",
                "Plant resistant apple varieties"
            ],
            "Powdery_mildew": [
                "Apply sulfur or potassium bicarbonate",
                "Reduce humidity around plants",
                "Use resistant varieties when available",
                "Prune affected areas to improve air circulation"
            ],
            "Cercospora_leaf_spot_Gray_leaf_spot": [
                "Apply fungicides containing chlorothalonil or copper",
                "Rotate crops to non-host plants",
                "Remove and destroy infected plant debris"
            ],
            "Common_rust": [
                "Use resistant varieties",
                "Apply fungicides if severe, such as those with propiconazole or azoxystrobin",
                "Practice good sanitation to remove rust spores"
            ],
            "Northern_Leaf_Blight": [
                "Use resistant varieties",
                "Apply fungicides if severe, typically those containing strobilurins or triazoles",
                "Practice crop rotation and residue management"
            ],
            "Esca_Black_Measles": [
                "Prune out diseased wood during dry periods",
                "Apply protective fungicidal paints to pruning wounds",
                "Improve vineyard hygiene to reduce inoculum"
            ],
            "Leaf_blight": [
                "Apply fungicides (e.g., copper-based or mancozeb)",
                "Improve air circulation around plants by proper spacing and pruning",
                "Remove and dispose of infected leaves"
            ],
            "Haunglongbing_Citrus_greening": [
                "No known cure; remove infected trees to prevent spread",
                "Control citrus psyllid vectors with insecticides",
                "Plant certified disease-free nursery stock"
            ],
            "Bacterial_spot": [
                "Apply copper-based bactericides",
                "Use disease-free seeds and transplants",
                "Avoid overhead watering and splashing soil onto plants",
                "Practice crop rotation"
            ],
            "Early_blight": [
                "Apply copper-based fungicides",
                "Practice crop rotation",
                "Remove and destroy infected plant material"
            ],
            "Late_blight": [
                "Apply fungicides containing chlorothalonil or copper",
                "Destroy infected plants immediately",
                "Avoid overhead watering"
            ],
            "Leaf_Mold": [
                "Improve air circulation by pruning and spacing plants",
                "Reduce humidity in greenhouses",
                "Apply fungicides if necessary, such as chlorothalonil or mancozeb"
            ],
            "Septoria_leaf_spot": [
                "Apply fungicides with chlorothalonil or mancozeb",
                "Remove infected leaves and plant debris",
                "Avoid overhead watering to minimize leaf wetness"
            ],
            "Spider_mites _Two_spotted_spider_mite": [
                "Apply insecticidal soaps or horticultural oils",
                "Increase humidity around plants (mist foliage)",
                "Introduce predatory mites"
            ],
            "Target_Spot": [
                "Apply fungicides containing chlorothalonil or mancozeb",
                "Practice crop rotation and sanitation",
                "Ensure good air circulation"
            ],
            "Tomato_Yellow_Leaf_Curl_Virus": [
                "No direct chemical cure for the virus; manage whitefly vectors with insecticides",
                "Use resistant varieties if available",
                "Remove infected plants immediately"
            ],
            "Tomato_mosaic_virus": [
                "No chemical cure; remove and destroy infected plants",
                "Disinfect tools and hands after handling infected plants",
                "Use resistant varieties or certified disease-free seeds"
            ],
            "healthy": [
                "Continue regular monitoring",
                "Maintain good cultural practices (proper watering, fertilization, pruning)",
                "Preventative measures like ensuring good air circulation and balanced nutrition"
            ]
        }

        try:
            disease_full_name = prediction.split("Disease:")[1].split("\n")[0].strip()
        except IndexError:
            disease_full_name = "unknown"

        treatment_suggestions = []

        found_treatment = False
        for key, suggestions in treatments.items():
            if key.replace("_", " ").lower() in disease_full_name.replace("_", " ").lower():
                treatment_suggestions = suggestions
                found_treatment = True
                break
    
        if not found_treatment and "healthy" in disease_full_name.lower():
            treatment_suggestions = treatments.get("healthy", ["No specific treatment needed. Maintain good plant health practices."])
            found_treatment = True

        if not found_treatment or not treatment_suggestions:
            treatment_suggestions = ["No specific treatment suggestions available for this diagnosis. Please consult an agricultural expert."]
    
        return treatment_suggestions


    def display_image(self, img, is_original=False):
        self.image_canvas.delete("all")
        
        if not is_original and self.uploaded_image:
            self.original_img = self.uploaded_image.copy()
        
        canvas_width = self.image_canvas.winfo_width()
        canvas_height = self.image_canvas.winfo_height()
        
        img_ratio = img.width / img.height
        canvas_ratio = canvas_width / canvas_height
        
        if img_ratio > canvas_ratio:
            new_width = min(img.width, canvas_width)
            new_height = int(new_width / img_ratio)
        else:
            new_height = min(img.height, canvas_height)
            new_width = int(new_height * img_ratio)
        
        img = img.resize((new_width, new_height), Image.Resampling.LANCZOS)
        
        self.current_display_img = ImageTk.PhotoImage(img)
        
        self.image_canvas.create_image(
            canvas_width//2,
            canvas_height//2,
            image=self.current_display_img,
            anchor=tk.CENTER,
            tags="image"
        )
        
        if not is_original:
            self.zoom_level = 1.0
            self.zoom_label.config(text=f"Zoom: {self.zoom_level:.1f}x")
        
        self.update_scrollregion()
        
        self.current_display_img_reference = self.current_display_img

    def zoom_image(self, event):
        if not hasattr(self, 'filtered_image') or not self.filtered_image:
            return
            
        scale = 1.0
        if event.delta > 0:
            scale *= 1.2
        else:
            scale /= 1.2
        
        self.zoom_level *= scale
        self.zoom_label.config(text=f"Zoom: {self.zoom_level:.1f}x")
        
        if self.zoom_level < 0.1:
            self.zoom_level = 0.1
        elif self.zoom_level > 10:
            self.zoom_level = 10
        
        new_width = int(self.filtered_image.width * self.zoom_level)
        new_height = int(self.filtered_image.height * self.zoom_level)
        
        img = self.filtered_image.resize((new_width, new_height), Image.Resampling.LANCZOS)
        self.current_display_img = ImageTk.PhotoImage(img)
        
        self.image_canvas.delete("all")
        self.image_canvas.create_image(
            self.image_canvas.winfo_width()//2,
            self.image_canvas.winfo_height()//2,
            image=self.current_display_img,
            anchor=tk.CENTER,
            tags="image"
        )
        
        self.current_display_img_reference = self.current_display_img
        self.update_scrollregion()

    def start_pan(self, event):
        self.pan_start_x = event.x
        self.pan_start_y = event.y
        self.image_canvas.scan_mark(event.x, event.y)

    def pan_image(self, event):
        self.image_canvas.scan_dragto(event.x, event.y, gain=1)

    def update_scrollregion(self, event=None):
        bbox = self.image_canvas.bbox("all")
        if bbox:
            self.image_canvas.configure(scrollregion=bbox)

    def activate_crop_mode(self):
        if not self.uploaded_image:
            self.controller._show_notification("Please upload an image first to crop.", "warning")
            return

        self.cropping_active = True
        self.status_label.config(text="Draw a rectangle to crop the image.")
        self.crop_button.config(state=tk.DISABLED)
        # Show apply and reset buttons
        self.apply_crop_button.pack(side=tk.LEFT, padx=5, pady=5)
        self.reset_crop_button.pack(side=tk.LEFT, padx=5, pady=5)

        self.image_canvas.unbind("<ButtonPress-1>")
        self.image_canvas.unbind("<B1-Motion>")
        self.image_canvas.bind("<ButtonPress-1>", self.start_crop)
        self.image_canvas.bind("<B1-Motion>", self.drag_crop)
        self.image_canvas.bind("<ButtonRelease-1>", self.end_crop)

        if self.crop_rect_id:
            self.image_canvas.delete(self.crop_rect_id)
            self.crop_rect_id = None
        self.cropped_image = None
        self.controller._show_notification("Crop mode activated. Draw a rectangle.", "info")


    def start_crop(self, event):
        self.start_x = self.image_canvas.canvasx(event.x)
        self.start_y = self.image_canvas.canvasy(event.y)
        if self.crop_rect_id:
            self.image_canvas.delete(self.crop_rect_id)
        self.crop_rect_id = self.image_canvas.create_rectangle(
            self.start_x, self.start_y, self.start_x, self.start_y,
            outline="red", width=2, dash=(5, 2)
        )

    def drag_crop(self, event):
        cur_x = self.image_canvas.canvasx(event.x)
        cur_y = self.image_canvas.canvasy(event.y)
        self.image_canvas.coords(self.crop_rect_id, self.start_x, self.start_y, cur_x, cur_y)

    def end_crop(self, event):
        self.apply_crop_button.config(state=tk.NORMAL)

    def apply_crop(self):
        if not self.crop_rect_id or not self.uploaded_image:
            self.controller._show_notification("Please draw a crop rectangle first.", "warning")
            return

        x1, y1, x2, y2 = self.image_canvas.coords(self.crop_rect_id)

        img_item = self.image_canvas.find_withtag("image")
        if not img_item:
            self.controller._show_notification("Image not found on canvas.", "error")
            return

        img_x, img_y = self.image_canvas.coords(img_item[0])
        
        displayed_img_width = self.current_display_img.width()
        displayed_img_height = self.current_display_img.height()

        img_display_x1 = img_x - displayed_img_width / 2
        img_display_y1 = img_y - displayed_img_height / 2

        crop_x1_rel = max(0, x1 - img_display_x1)
        crop_y1_rel = max(0, y1 - img_display_y1)
        crop_x2_rel = min(displayed_img_width, x2 - img_display_x1)
        crop_y2_rel = min(displayed_img_height, y2 - img_display_y1)

        original_width, original_height = self.uploaded_image.size
        
        scale_x = original_width / displayed_img_width
        scale_y = original_height / displayed_img_height

        final_x1 = int(crop_x1_rel * scale_x)
        final_y1 = int(crop_y1_rel * scale_y)
        final_x2 = int(crop_x2_rel * scale_x)
        final_y2 = int(crop_y2_rel * scale_y)

        if final_x2 <= final_x1 or final_y2 <= final_y1:
            self.controller._show_notification("Please draw a valid crop area.", "warning")
            return

        self.cropped_image = self.uploaded_image.crop((final_x1, final_y1, final_x2, final_y2))
        self.filtered_image = self.cropped_image.copy()
        self.display_image(self.cropped_image)
        self.status_label.config(text="Image cropped. Ready for diagnosis.")
        
        self.cropping_active = False
        self.image_canvas.unbind("<ButtonPress-1>")
        self.image_canvas.unbind("<B1-Motion>")
        self.image_canvas.unbind("<ButtonRelease-1>")
        self.image_canvas.bind("<ButtonPress-1>", self.start_pan)
        self.image_canvas.bind("<B1-Motion>", self.pan_image)
        self.image_canvas.delete(self.crop_rect_id)
        self.crop_rect_id = None
        self.crop_button.config(state=tk.NORMAL)
        self.apply_crop_button.pack_forget()
        self.reset_crop_button.pack_forget() # Hide reset button after applying crop

        self.start_processing(self.current_image_path)
        self.controller._show_notification("Image cropped successfully.", "success")


    def reset_crop(self):
        if self.original_img:
            self.uploaded_image = self.original_img.copy()
            self.filtered_image = self.original_img.copy()
            self.cropped_image = None
            self.display_image(self.uploaded_image)
            self.status_label.config(text="Crop reset. Original image restored.")
            
            self.apply_crop_button.pack_forget()
            self.reset_crop_button.pack_forget()
            self.crop_button.config(state=tk.NORMAL)

            self.cropping_active = False
            self.image_canvas.unbind("<ButtonPress-1>")
            self.image_canvas.unbind("<B1-Motion>")
            self.image_canvas.unbind("<ButtonRelease-1>")
            self.image_canvas.bind("<ButtonPress-1>", self.start_pan)
            self.image_canvas.bind("<B1-Motion>", self.pan_image)

            self.start_processing(self.current_image_path)
            self.controller._show_notification("Crop reset. Original image restored.", "info")


    def create_filter_controls(self, parent):
        filter_frame = ttk.LabelFrame(parent, text="Image Enhancement Tools", style='Section.TLabelframe')
        filter_frame.grid(row=5, column=0, columnspan=2, pady=10, padx=10, sticky="ew") # Placed below main content area
        
        self.filter_notebook = ttk.Notebook(filter_frame)
        self.filter_notebook.pack(fill='both', expand=True)
        
        basic_frame = ttk.Frame(self.filter_notebook)
        self.filter_notebook.add(basic_frame, text="Basic Adjustments")
        
        # Configure grid for basic_frame
        basic_frame.grid_columnconfigure(1, weight=1) # Allow scale to expand

        ttk.Label(basic_frame, text="Brightness:").grid(row=0, column=0, padx=5, pady=5, sticky="e")
        self.brightness_scale = ttk.Scale(
            basic_frame,
            from_=0.1,
            to=2.0,
            value=1.0,
            command=lambda v: (self.adjust_brightness(float(v)), self.brightness_value_label.config(text=f"{float(v):.1f}")))
        self.brightness_scale.grid(row=0, column=1, padx=5, pady=5, sticky="ew")
        self.brightness_value_label = ttk.Label(basic_frame, text="1.0")
        self.brightness_value_label.grid(row=0, column=2, padx=(0,5))
        
        ttk.Label(basic_frame, text="Contrast:").grid(row=1, column=0, padx=5, pady=5, sticky="e")
        self.contrast_scale = ttk.Scale(
            basic_frame,
            from_=0.1,
            to=2.0,
            value=1.0,
            command=lambda v: (self.adjust_contrast(float(v)), self.contrast_value_label.config(text=f"{float(v):.1f}")))
        self.contrast_scale.grid(row=1, column=1, padx=5, pady=5, sticky="ew")
        self.contrast_value_label = ttk.Label(basic_frame, text="1.0")
        self.contrast_value_label.grid(row=1, column=2, padx=(0,5))
        
        ttk.Label(basic_frame, text="Saturation:").grid(row=2, column=0, padx=5, pady=5, sticky="e")
        self.saturation_scale = ttk.Scale(
            basic_frame,
            from_=0.0,
            to=2.0,
            value=1.0,
            command=lambda v: (self.adjust_saturation(float(v)), self.saturation_value_label.config(text=f"{float(v):.1f}")))
        self.saturation_scale.grid(row=2, column=1, padx=5, pady=5, sticky="ew")
        self.saturation_value_label = ttk.Label(basic_frame, text="1.0")
        self.saturation_value_label.grid(row=2, column=2, padx=(0,5))
        
        advanced_frame = ttk.Frame(self.filter_notebook)
        self.filter_notebook.add(advanced_frame, text="Advanced Adjustments")

        # Configure grid for advanced_frame
        advanced_frame.grid_columnconfigure(1, weight=1) # Allow scale to expand
        
        ttk.Label(advanced_frame, text="Edge Enhancement:").grid(row=0, column=0, padx=5, pady=5, sticky="e")
        self.edge_scale = ttk.Scale(
            advanced_frame,
            from_=0,
            to=2,
            value=0,
            command=lambda v: (self.enhance_edges(float(v)), self.edge_value_label.config(text=f"{float(v):.1f}")))
        self.edge_scale.grid(row=0, column=1, padx=5, pady=5, sticky="ew")
        self.edge_value_label = ttk.Label(advanced_frame, text="0.0")
        self.edge_value_label.grid(row=0, column=2, padx=(0,5))
        
        ttk.Label(advanced_frame, text="Blur Reduction:").grid(row=1, column=0, padx=5, pady=5, sticky="e")
        self.blur_scale = ttk.Scale(
            advanced_frame,
            from_=0,
            to=2,
            value=0,
            command=lambda v: (self.reduce_blur(float(v)), self.blur_value_label.config(text=f"{float(v):.1f}")))
        self.blur_scale.grid(row=1, column=1, padx=5, pady=5, sticky="ew")
        self.blur_value_label = ttk.Label(advanced_frame, text="0.0")
        self.blur_value_label.grid(row=1, column=2, padx=(0,5))
        
        ttk.Label(advanced_frame, text="Color Balance:").grid(row=2, column=0, padx=5, pady=5, sticky="e")
        self.color_balance_scale = ttk.Scale(
            advanced_frame,
            from_=-1.0,
            to=1.0,
            value=0.0,
            command=lambda v: (self.adjust_color_balance(float(v)), self.color_balance_value_label.config(text=f"{float(v):.1f}")))
        self.color_balance_scale.grid(row=2, column=1, padx=5, pady=5, sticky="ew")
        self.color_balance_value_label = ttk.Label(advanced_frame, text="0.0")
        self.color_balance_value_label.grid(row=2, column=2, padx=(0,5))
        
        presets_frame = ttk.Frame(filter_frame)
        presets_frame.pack(fill='x', pady=(5,0))
        
        presets = [
            ("Auto Contrast", self.auto_contrast),
            ("Sharpen", self.sharpen_image),
            ("Grayscale", self.grayscale_image),
            ("Reset", self.reset_filters)
        ]
        
        for i, (text, command) in enumerate(presets):
            ttk.Button(
                presets_frame,
                text=text,
                command=command,
                style='TButton'
            ).grid(row=0, column=i, padx=2, sticky="ew")
            presets_frame.grid_columnconfigure(i, weight=1)

    def adjust_brightness(self, factor):
        source_image = self.cropped_image if self.cropped_image else self.uploaded_image
        if source_image and self.preview_var.get():
            enhancer = ImageEnhance.Brightness(source_image)
            self.filtered_image = enhancer.enhance(factor)
            self.display_image(self.filtered_image)

    def adjust_contrast(self, factor):
        source_image = self.cropped_image if self.cropped_image else self.uploaded_image
        if source_image and self.preview_var.get():
            enhancer = ImageEnhance.Contrast(source_image)
            self.filtered_image = enhancer.enhance(factor)
            self.display_image(self.filtered_image)

    def adjust_saturation(self, factor):
        source_image = self.cropped_image if self.cropped_image else self.uploaded_image
        if source_image and self.preview_var.get():
            enhancer = ImageEnhance.Color(source_image)
            self.filtered_image = enhancer.enhance(factor)
            self.display_image(self.filtered_image)

    def enhance_edges(self, factor):
        source_image = self.cropped_image if self.cropped_image else self.uploaded_image
        if source_image and self.preview_var.get():
            if factor > 0:
                self.filtered_image = source_image.filter(
                    ImageFilter.UnsharpMask(radius=2, percent=150, threshold=3))
                enhancer = ImageEnhance.Sharpness(self.filtered_image)
                self.filtered_image = enhancer.enhance(1.0 + factor)
            else:
                self.filtered_image = source_image.copy()
            self.display_image(self.filtered_image)

    def reduce_blur(self, factor):
        source_image = self.cropped_image if self.cropped_image else self.uploaded_image
        if source_image and self.preview_var.get():
            if factor > 0:
                self.filtered_image = source_image.filter(
                    ImageFilter.MedianFilter(size=3))
                enhancer = ImageEnhance.Sharpness(self.filtered_image)
                self.filtered_image = enhancer.enhance(1.0 + factor)
            else:
                self.filtered_image = source_image.copy()
            self.display_image(self.filtered_image)

    def adjust_color_balance(self, factor):
        source_image = self.cropped_image if self.cropped_image else self.uploaded_image
        if source_image and self.preview_var.get():
            img_array = np.array(source_image)
            
            if factor > 0:
                img_array[:,:,0] = np.clip(img_array[:,:,0] * (1 + factor), 0, 255)
                img_array[:,:,1] = np.clip(img_array[:,:,1] * (1 + factor/2), 0, 255)
            else:
                img_array[:,:,2] = np.clip(img_array[:,:,2] * (1 - factor), 0, 255)
            
            self.filtered_image = Image.fromarray(img_array)
            self.display_image(self.filtered_image)

    def auto_contrast(self):
        source_image = self.cropped_image if self.cropped_image else self.uploaded_image
        if source_image:
            if source_image.mode == "RGBA":
                source_image = source_image.convert("RGB")
            self.filtered_image = ImageOps.autocontrast(source_image)
            self.display_image(self.filtered_image)
            self.brightness_scale.set(1.0)
            self.contrast_scale.set(1.0)
            self.brightness_value_label.config(text="1.0")
            self.contrast_value_label.config(text="1.0")
            self.controller._show_notification("Auto contrast applied.", "info")


    def sharpen_image(self):
        source_image = self.cropped_image if self.cropped_image else self.uploaded_image
        if source_image:
            self.filtered_image = source_image.filter(ImageFilter.SHARPEN)
            self.display_image(self.filtered_image)
            self.controller._show_notification("Image sharpened.", "info")


    def grayscale_image(self):
        source_image = self.cropped_image if self.cropped_image else self.uploaded_image
        if source_image:
            self.filtered_image = source_image.convert("L").convert("RGB")
            self.display_image(self.filtered_image)
            self.saturation_scale.set(0.0)
            self.saturation_value_label.config(text="0.0")
            self.controller._show_notification("Image converted to grayscale.", "info")


    def reset_filters(self):
        source_image = self.cropped_image if self.cropped_image else self.uploaded_image
        if source_image:
            self.filtered_image = source_image.copy()
            self.display_image(self.filtered_image)
            self.brightness_scale.set(1.0)
            self.contrast_scale.set(1.0)
            self.saturation_scale.set(1.0)
            self.edge_scale.set(0.0)
            self.blur_scale.set(0.0)
            self.color_balance_scale.set(0.0)
            self.brightness_value_label.config(text="1.0")
            self.contrast_value_label.config(text="1.0")
            self.saturation_value_label.config(text="1.0")
            self.edge_value_label.config(text="0.0")
            self.blur_value_label.config(text="0.0")
            self.color_balance_value_label.config(text="0.0")
            self.controller._show_notification("Image filters reset.", "info")


    def toggle_live_preview(self):
        if self.preview_var.get() and self.filtered_image:
            self.display_image(self.filtered_image)
            self.controller._show_notification("Live preview enabled.", "info")
        else:
            self.controller._show_notification("Live preview disabled.", "info")


    def clear_results(self):
        self.result_text_area.config(state=tk.NORMAL)
        self.result_text_area.delete(1.0, tk.END)
        self.result_text_area.insert(tk.END, "Processing image...")
        self.result_text_area.config(state=tk.DISABLED)

    def diagnose_disease(self, image_path):
        if self.controller.disease_model is None:
            return "Error: Disease detection model not loaded"
        
        is_non_crop = self.controller.is_non_crop_image(image_path)
        if is_non_crop:
            return "Error: Uploaded image doesn't appear to be a crop leaf image"
        
        try:
            img = image.load_img(image_path, target_size=(150, 150))
            img_array = image.img_to_array(img) / 255.0
            img_array = np.expand_dims(img_array, axis=0)

            predictions = self.controller.disease_model.predict(img_array, verbose=0)

            predicted_class = np.argmax(predictions, axis=1)[0]
            confidence = np.max(predictions) * 100
            
            predicted_label = list(self.controller.class_labels.keys())[predicted_class]
            disease_name = " ".join(predicted_label.split("_")[1:])
            crop_type = predicted_label.split("_")[0]

            if confidence < 50:
                result = (
                    f"Crop Type: {crop_type}\n"
                    f"Disease: {disease_name}\n"
                    f"Confidence: {confidence:.1f}%\n"
                    "Note: Low confidence prediction\n"
                    "Please upload a clearer image for better results"
                )
            else:
                result = (
                    f"Crop Type: {crop_type}\n"
                    f"Disease: {disease_name}\n"
                    f"Confidence: {confidence:.1f}%"
                )
            return result
            
        except Exception as e:
            logging.error(f"Diagnosis error in diagnose_disease: {e}", exc_info=True)
            return f"Error during diagnosis: {str(e)}"

    def print_current_report(self):
        if not self.current_image_path or not self.filtered_image:
            self.controller._show_notification("No image or diagnosis available to print.", "warning")
            return
        
        temp_img_path = None
        temp_pdf_path = None # Initialize temp_pdf_path
        try:
            pdf = FPDF()
            pdf.set_font("Helvetica", "B", 16) # Changed font to Helvetica
            pdf.add_page()
            
            pdf.cell(0, 10, "Crop Disease Diagnosis Report", ln=True, align="C")
            pdf.ln(10)
            
            pdf.set_font("Helvetica", size=12) # Changed font to Helvetica
            pdf.cell(0, 10, f"Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}", ln=True)
            pdf.cell(0, 10, f"Image: {os.path.basename(self.current_image_path)}", ln=True)
            pdf.ln(5)

            pdf.set_font("Helvetica", "B", 14) # Changed font to Helvetica
            pdf.cell(0, 10, "Diagnosis Results:", ln=True)
            pdf.set_font("Helvetica", size=12) # Changed font to Helvetica
            
            crop_type = self.last_diagnosis_details.get("crop_type", "N/A")
            disease_name = self.last_diagnosis_details.get("disease_name", "N/A")
            confidence = self.last_diagnosis_details.get("confidence", "N/A")
            pdf.cell(0, 7, f"Crop Type: {crop_type}", ln=True)
            pdf.cell(0, 7, f"Disease: {disease_name}", ln=True)
            pdf.cell(0, 7, f"Confidence: {confidence}", ln=True)
            pdf.ln(5)

            pdf.set_font("Helvetica", "B", 14) # Changed font to Helvetica
            pdf.cell(0, 10, "Disease Information:", ln=True)
            pdf.set_font("Helvetica", size=12) # Changed font to Helvetica
            full_disease_key = f"{crop_type}_{disease_name.replace(' ', '_')}"
            info = self.disease_info.get(full_disease_key, self.disease_info.get("healthy"))
            if info:
                pdf.multi_cell(0, 7, f"Description: {info['description']}")
                pdf.ln(2)
                pdf.multi_cell(0, 7, f"Symptoms: {info['symptoms']}")
            else:
                pdf.multi_cell(0, 7, "No detailed information available for this disease.")
            pdf.ln(10)

            pdf.set_font("Helvetica", "B", 14) # Changed font to Helvetica
            pdf.cell(0, 10, "Treatment Suggestions:", ln=True)
            pdf.set_font("Helvetica", size=12) # Changed font to Helvetica

            if self.last_treatment_suggestions:
                for suggestion in self.last_treatment_suggestions:
                    pdf.multi_cell(0, 7, f"- {suggestion}")
            else:
                pdf.multi_cell(0, 7, "No specific treatment suggestions available for this diagnosis.")
            pdf.ln(10)
            
            if self.filtered_image:
                try:
                    img_to_save = self.filtered_image.copy()
                    img_to_save = img_to_save.convert("RGB")

                    img_width, img_height = img_to_save.size
                    max_pdf_width = 150
                    max_pdf_height = 150

                    if img_width > max_pdf_width or img_height > max_pdf_height:
                        img_to_save.thumbnail((max_pdf_width, max_pdf_height), Image.Resampling.LANCZOS)
                    
                    with tempfile.NamedTemporaryFile(delete=False, suffix=".jpg") as temp_file:
                        temp_img_path = temp_file.name
                        img_to_save.save(temp_img_path, format="JPEG")

                    pdf.ln(5)
                    pdf.cell(0, 10, "Image Preview:", ln=True)
                    x_pos = (pdf.w - img_to_save.width) / 2
                    pdf.image(temp_img_path, x=x_pos, w=img_to_save.width)

                except Exception as img_embed_e:
                    logging.error(f"Failed to embed image in PDF for print: {img_embed_e}", exc_info=True)
                    pdf.set_font("Helvetica", "I", 10) # Changed font to Helvetica
                    pdf.cell(0, 10, "Error: Could not embed image in report. Details: " + str(img_embed_e), ln=True, align="C")
                finally:
                    if temp_img_path and os.path.exists(temp_img_path):
                        os.remove(temp_img_path)
            else:
                logging.warning("filtered_image was None when attempting to print report image.")
                pdf.set_font("Helvetica", "I", 10) # Changed font to Helvetica
                pdf.cell(0, 10, "No image available to embed in report.", ln=True, align="C")
            
            temp_pdf_path = os.path.join(tempfile.gettempdir(), "current_diagnosis_report.pdf")
            pdf.output(temp_pdf_path)

            try:
                webbrowser.open_new(temp_pdf_path)
                self.controller._show_notification("Report generated for printing. Check your PDF viewer.", "success", duration=5000)
            except Exception as web_e:
                self.controller._show_notification(f"Report generated, but failed to open in browser: {web_e}. Please open it manually from: {os.path.basename(temp_pdf_path)}", "warning", duration=7000)
                logging.error(f"Failed to open PDF in browser: {web_e}", exc_info=True)

        except Exception as e:
            self.controller._show_notification(f"Failed to generate printable report: {e}", "error", duration=5000)
            logging.error(f"Print report error: {e}", exc_info=True)
        finally:
            # Ensure temp_pdf_path is cleaned up after printing (if it was created)
            if temp_pdf_path and os.path.exists(temp_pdf_path):
                # Give a moment for the external PDF viewer to open before deleting
                threading.Timer(2, lambda: os.remove(temp_pdf_path)).start()


    def save_report(self):
        if not self.current_image_path or not self.filtered_image:
            self.controller._show_notification("No image or diagnosis available to generate report.", "warning")
            return
            
        crop_type = self.last_diagnosis_details.get("crop_type", "N/A")
        disease_name = self.last_diagnosis_details.get("disease_name", "N/A")
        confidence = self.last_diagnosis_details.get("confidence", "N/A")

        default_filename = (
            f"{datetime.now().strftime('%Y%m%d_%H%M%S')}_"
            f"{os.path.splitext(os.path.basename(self.current_image_path))[0]}_report.pdf"
        )
        
        file_path = filedialog.asksaveasfilename(
            defaultextension=".pdf",
            filetypes=[("PDF files", "*.pdf")],
            initialfile=default_filename,
            title="Save Diagnosis Report"
        )
        
        if not file_path: # User cancelled the save dialog
            self.controller._show_notification("Report saving cancelled.", "info")
            return

        temp_img_path = None
        try:
            pdf = FPDF()
            pdf.set_font("Helvetica", "B", 16) # Changed font to Helvetica
            pdf.add_page()
            
            pdf.cell(0, 10, "Crop Disease Diagnosis Report", ln=True, align="C")
            pdf.ln(10)
            
            pdf.set_font("Helvetica", size=12) # Changed font to Helvetica
            pdf.cell(0, 10, f"Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}", ln=True)
            pdf.cell(0, 10, f"Image: {os.path.basename(self.current_image_path)}", ln=True)
            pdf.ln(5)

            pdf.set_font("Helvetica", "B", 14) # Changed font to Helvetica
            pdf.cell(0, 10, "Diagnosis Results:", ln=True)
            pdf.set_font("Helvetica", size=12) # Changed font to Helvetica
            
            pdf.cell(0, 7, f"Crop Type: {crop_type}", ln=True)
            pdf.cell(0, 7, f"Disease: {disease_name}", ln=True)
            pdf.cell(0, 7, f"Confidence: {confidence}", ln=True)
            pdf.ln(5)

            pdf.set_font("Helvetica", "B", 14) # Changed font to Helvetica
            pdf.cell(0, 10, "Disease Information:", ln=True)
            pdf.set_font("Helvetica", size=12) # Changed font to Helvetica
            full_disease_key = f"{crop_type}_{disease_name.replace(' ', '_')}"
            info = self.disease_info.get(full_disease_key, self.disease_info.get("healthy"))
            if info:
                pdf.multi_cell(0, 7, f"Description: {info['description']}")
                pdf.ln(2)
                pdf.multi_cell(0, 7, f"Symptoms: {info['symptoms']}")
            else:
                pdf.multi_cell(0, 7, "No detailed information available for this disease.")
            pdf.ln(10)


            pdf.set_font("Helvetica", "B", 14) # Changed font to Helvetica
            pdf.cell(0, 10, "Treatment Suggestions:", ln=True)
            pdf.set_font("Helvetica", size=12) # Changed font to Helvetica

            if self.last_treatment_suggestions:
                for suggestion in self.last_treatment_suggestions:
                    pdf.multi_cell(0, 7, f"- {suggestion}")
            else:
                pdf.multi_cell(0, 7, "No specific treatment suggestions available for this diagnosis.")
            pdf.ln(10)
            
            if self.filtered_image:
                try:
                    img_to_save = self.filtered_image.copy()
                    img_to_save = img_to_save.convert("RGB")

                    img_width, img_height = img_to_save.size
                    max_pdf_width = 150
                    max_pdf_height = 150

                    if img_width > max_pdf_width or img_height > max_pdf_height:
                        img_to_save.thumbnail((max_pdf_width, max_pdf_height), Image.Resampling.LANCZOS)
                    
                    with tempfile.NamedTemporaryFile(delete=False, suffix=".jpg") as temp_file:
                        temp_img_path = temp_file.name
                        img_to_save.save(temp_img_path, format="JPEG")

                    pdf.ln(5)
                    pdf.cell(0, 10, "Image Preview:", ln=True)
                    x_pos = (pdf.w - img_to_save.width) / 2
                    pdf.image(temp_img_path, x=x_pos, w=img_to_save.width)

                except Exception as img_embed_e:
                    logging.error(f"Failed to embed image in PDF: {img_embed_e}", exc_info=True)
                    pdf.set_font("Helvetica", "I", 10) # Changed font to Helvetica
                    pdf.cell(0, 10, "Error: Could not embed image in report. Details: " + str(img_embed_e), ln=True, align="C")
                finally:
                    if temp_img_path and os.path.exists(temp_img_path):
                        os.remove(temp_img_path)
            else:
                logging.warning("filtered_image was None when attempting to save report image.")
                pdf.set_font("Helvetica", "I", 10) # Changed font to Helvetica
                pdf.cell(0, 10, "No image available to embed in report.", ln=True, align="C")
            
            pdf.output(file_path) # This is the critical line for saving
            self.controller._show_notification(f"Report saved to: {os.path.basename(file_path)}", "success", duration=5000)
            
        except Exception as e:
            self.controller._show_notification(f"Failed to save PDF: {e}", "error", duration=5000)
            logging.error(f"PDF save error: {e}", exc_info=True)            


class HistoryPage(ttk.Frame):
    def __init__(self, parent, controller):
        ttk.Frame.__init__(self, parent)
        self.controller = controller
        self.selected_item_data = None # Store the full data of the selected item
        self.history_view = None
        self.sort_order = "recent"
        self.filter_crop = "all"

        self.grid_rowconfigure(1, weight=1)
        self.grid_columnconfigure(0, weight=1)

        control_frame = ttk.Frame(self)
        control_frame.grid(row=0, column=0, columnspan=4, pady=(10, 5), padx=10, sticky="ew")
        
        title_frame = ttk.Frame(control_frame)
        title_frame.pack(side=tk.LEFT, padx=(0, 15))
            
        title_label = ttk.Label(
            title_frame, 
            text="Diagnosis History", 
            font=("Inter", 16, "bold")
        )
        title_label.pack(side=tk.LEFT)

        filter_frame = ttk.Frame(control_frame)
        filter_frame.pack(side=tk.RIGHT, fill=tk.Y)
        
        ttk.Label(filter_frame, text="Sort by:").pack(side=tk.LEFT, padx=(0, 5))
        self.sort_var = tk.StringVar(value="Most Recent")
        sort_menu = ttk.OptionMenu(
            filter_frame,
            self.sort_var,
            "Most Recent",
            "Most Recent",
            "Oldest First",
            "Highest Confidence",
            "Lowest Confidence",
            command=self.refresh_history
        )
        sort_menu.pack(side=tk.LEFT, padx=(0, 10))

        ttk.Label(filter_frame, text="Filter by crop:").pack(side=tk.LEFT, padx=(0, 5))
        self.filter_var = tk.StringVar(value="All Crops")

        crop_types = set()
        if self.controller.class_labels:
            for label in self.controller.class_labels.keys():
                crop_types.add(label.split('_')[0])
        
        sorted_crop_types = sorted(list(crop_types))
        
        menu_options = ["All Crops"] + sorted_crop_types

        crop_menu = ttk.OptionMenu(
            filter_frame,
            self.filter_var,
            *menu_options,
            command=self.refresh_history
        )
        crop_menu.pack(side=tk.LEFT)

        content_frame = ttk.Frame(self)
        content_frame.grid(row=1, column=0, sticky="nsew", padx=10, pady=5)
        content_frame.grid_columnconfigure(0, weight=1, minsize=300)
        content_frame.grid_columnconfigure(1, weight=2)

        list_frame = ttk.LabelFrame(content_frame, text="Recent Diagnoses")
        list_frame.grid(row=0, column=0, sticky="nsew", padx=(0, 5), pady=5)
        list_frame.grid_rowconfigure(0, weight=1)
        list_frame.grid_columnconfigure(0, weight=1)
        
        self.history_list = ttk.Treeview(
            list_frame,
            columns=("date", "crop", "disease", "confidence"),
            show="headings",
            selectmode="browse",
            height=15
        )
        
        self.history_list.heading("date", text="Date", anchor=tk.W)
        self.history_list.heading("crop", text="Crop", anchor=tk.W)
        self.history_list.heading("disease", text="Disease", anchor=tk.W)
        self.history_list.heading("confidence", text="Confidence", anchor=tk.W)
        
        self.history_list.column("date", width=120, stretch=False)
        self.history_list.column("crop", width=100, stretch=False)
        self.history_list.column("disease", width=150)
        self.history_list.column("confidence", width=100, stretch=False)
        
        scrollbar = ttk.Scrollbar(list_frame, orient=tk.VERTICAL, command=self.history_list.yview)
        self.history_list.configure(yscrollcommand=scrollbar.set)
        
        self.history_list.grid(row=0, column=0, sticky="nsew")
        scrollbar.grid(row=0, column=1, sticky="ns")
        
        self.history_list.bind("<<TreeviewSelect>>", self.on_history_select)

        detail_frame = ttk.LabelFrame(content_frame, text="Diagnosis Details")
        detail_frame.grid(row=0, column=1, sticky="nsew", padx=(5, 0), pady=5)
        detail_frame.grid_rowconfigure(1, weight=1)
        detail_frame.grid_columnconfigure(0, weight=1)
        
        self.image_preview = ttk.Label(detail_frame)
        self.image_preview.grid(row=0, column=0, pady=(5, 10))
        
        self.details_text = tk.Text(
            detail_frame,
            wrap=tk.WORD,
            font=("Inter", 10),
            padx=5,
            pady=5,
            height=10,
            state=tk.DISABLED
        )
        # Configure tags for structured display
        self.details_text.tag_configure("header", font=("Inter", 12, "bold"), foreground=self.controller.colors['fg'])
        self.details_text.tag_configure("disease_icon", font=("Arial", 12))
        self.details_text.tag_configure("confidence_high", foreground="green")
        self.details_text.tag_configure("confidence_medium", foreground="orange")
        self.details_text.tag_configure("confidence_low", foreground="red")
        self.details_text.tag_configure("separator", foreground="#CCCCCC")
        self.details_text.tag_configure("info_header", font=("Inter", 11, "bold"), foreground="#007BFF")
        self.details_text.tag_configure("info_content", lmargin1=20, lmargin2=20)
        self.details_text.tag_configure("treatment_header", font=("Inter", 11, "bold"), foreground="#6C757D")
        self.details_text.tag_configure("treatment_bullet", lmargin1=20, lmargin2=20)

        scrollbar_text = ttk.Scrollbar(detail_frame, command=self.details_text.yview)
        self.details_text.configure(yscrollcommand=scrollbar_text.set) # Corrected: use scrollbar_text.set
        
        self.details_text.grid(row=1, column=0, sticky="nsew")
        scrollbar_text.grid(row=1, column=1, sticky="ns")

        # Buttons for selected item actions
        selected_item_buttons_frame = ttk.Frame(detail_frame)
        selected_item_buttons_frame.grid(row=2, column=0, columnspan=2, pady=(10, 0), sticky="ew")
        selected_item_buttons_frame.grid_columnconfigure((0,1,2,3), weight=1)

        self.view_details_button = ttk.Button(
            selected_item_buttons_frame,
            text="View Details",
            command=self.view_selected_details,
            state=tk.DISABLED,
            style='TButton'
        )
        self.view_details_button.grid(row=0, column=0, padx=2, pady=2, sticky="ew")

        self.view_image_button = ttk.Button(
            selected_item_buttons_frame,
            text="View Image",
            command=self.view_full_image,
            state=tk.DISABLED,
            style='TButton'
        )
        self.view_image_button.grid(row=0, column=1, padx=2, pady=2, sticky="ew")

        self.re_diagnose_button = ttk.Button(
            selected_item_buttons_frame,
            text="Re-diagnose",
            command=self.re_diagnose_image,
            state=tk.DISABLED,
            style='TButton'
        )
        self.re_diagnose_button.grid(row=0, column=2, padx=2, pady=2, sticky="ew")
        
        self.delete_selected_button = ttk.Button(
            selected_item_buttons_frame,
            text="Delete Selected",
            command=self.delete_selected,
            state=tk.DISABLED,
            style='TButton'
        )
        self.delete_selected_button.grid(row=0, column=3, padx=2, pady=2, sticky="ew")


        # Global history management buttons
        button_frame = ttk.Frame(self)
        button_frame.grid(row=2, column=0, columnspan=2, pady=10, padx=10, sticky="ew")
        
        buttons = [
            ("Export All History", self.export_all_history),
            ("Import History", self.import_history),
            ("Clear All History", self.clear_history),
            ("Back to Home", lambda: controller.show_frame("HomePage"))
        ]
        
        for i, (text, command) in enumerate(buttons):
            btn = ttk.Button(
                button_frame,
                text=text,
                command=command,
                style='TButton'
            )
            btn.grid(row=0, column=i, padx=5, pady=5, sticky="ew")
            button_frame.grid_columnconfigure(i, weight=1)

        self.refresh_history()
        self.clear_details_display() # Clear details on initial load

    def on_show(self):
        """Called when the HistoryPage is brought to the front."""
        self.refresh_history()
        self.clear_details_display() # Ensure details are cleared when navigating back to this page

    def refresh_history(self, *args):
        self.history_list.delete(*self.history_list.get_children())
        self.selected_item_data = None # Clear selected item data
        self.clear_details_display() # Clear details display

        history = self.get_filtered_history()

        for item in history:
            # Use 'details' if available, otherwise parse from 'diagnosis' string
            if "details" in item and item["details"]:
                crop = item["details"].get("crop_type", "Unknown")
                disease = item["details"].get("disease_name", "Unknown")
                confidence = item["details"].get("confidence", "N/A")
            else:
                # Fallback for older entries or if 'details' was not saved
                crop = "Unknown"
                disease = "Unknown"
                confidence = self.extract_confidence(item["diagnosis"])
                confidence = f"{confidence:.1f}%" if confidence is not None else "N/A"

            self.history_list.insert("", "end", 
                values=(
                    item["timestamp"],
                    crop,
                    disease,
                    confidence
                ),
                tags=("healthy" if "healthy" in disease.lower() else ""),
                iid=item["timestamp"] # Use timestamp as unique ID for easy retrieval
            )
        
        self.history_list.tag_configure("healthy", foreground="green")

    def get_filtered_history(self):
        history = self.controller.diagnosis_history.copy()
        
        crop_filter = self.filter_var.get().lower()
        if crop_filter != "all crops":
            history = [h for h in history if h.get("details", {}).get("crop_type", "").lower() == crop_filter]
        
        sort_key = self.sort_var.get().lower()
        if sort_key == "oldest first":
            history.sort(key=lambda x: x["timestamp"])
        elif sort_key == "highest confidence":
            history.sort(key=lambda x: float(x.get("details", {}).get("confidence", "0%").replace("%", "")) if x.get("details", {}).get("confidence") else 0, reverse=True)
        elif sort_key == "lowest confidence":
            history.sort(key=lambda x: float(x.get("details", {}).get("confidence", "0%").replace("%", "")) if x.get("details", {}).get("confidence") else 0)
        else: # Most Recent
            history.sort(key=lambda x: x["timestamp"], reverse=True)
            
        return history

    def extract_confidence(self, diagnosis_text):
        if not diagnosis_text:
            return None
            
        import re
        match = re.search(r"Confidence:\s*([\d.]+)%", diagnosis_text)
        return float(match.group(1)) if match else None

    def on_history_select(self, event):
        selected_item_id = self.history_list.selection()
        if not selected_item_id:
            self.selected_item_data = None
            self.clear_details_display()
            # Disable buttons if no item is selected
            self.view_details_button.config(state=tk.DISABLED)
            self.view_image_button.config(state=tk.DISABLED)
            self.re_diagnose_button.config(state=tk.DISABLED)
            self.delete_selected_button.config(state=tk.DISABLED)
            # self.export_selected_button.config(state=tk.DISABLED) # This button is now part of the global buttons, will manage separately if needed.
            return
            
        # Retrieve the full item data using its unique ID (timestamp)
        selected_timestamp = selected_item_id[0]
        self.selected_item_data = next((item for item in self.controller.diagnosis_history if item["timestamp"] == selected_timestamp), None)
        
        if self.selected_item_data:
            # Enable buttons when an item is selected
            self.view_details_button.config(state=tk.NORMAL)
            self.view_image_button.config(state=tk.NORMAL if self.selected_item_data.get("image") and os.path.exists(self.selected_item_data["image"]) else tk.DISABLED)
            self.re_diagnose_button.config(state=tk.NORMAL if self.selected_item_data.get("image") and os.path.exists(self.selected_item_data["image"]) else tk.DISABLED)
            self.delete_selected_button.config(state=tk.NORMAL)
            # self.export_selected_button.config(state=tk.NORMAL) # This button is now part of the global buttons, will manage separately if needed.
            self.show_history_image(self.selected_item_data.get("image"))
            self.view_selected_details() # Automatically show details on selection
        else:
            self.clear_details_display()
            self.view_details_button.config(state=tk.DISABLED)
            self.view_image_button.config(state=tk.DISABLED)
            self.re_diagnose_button.config(state=tk.DISABLED)
            self.delete_selected_button.config(state=tk.DISABLED)


    def clear_details_display(self):
        """Clears the details text area and image preview."""
        self.details_text.config(state=tk.NORMAL)
        self.details_text.delete(1.0, tk.END)
        self.details_text.insert(tk.END, "Select a diagnosis from the list to view its details.")
        self.details_text.config(state=tk.DISABLED)
        self.image_preview.config(image="")
        self.image_preview.image = None
        self.selected_item_data = None # Ensure no item is considered selected


    def view_selected_details(self):
        if not self.selected_item_data:
            self.controller._show_notification("No item selected to view details.", "warning")
            return

        item = self.selected_item_data
        self.details_text.config(state=tk.NORMAL)
        self.details_text.delete(1.0, tk.END)

        # Display basic diagnosis info
        self.details_text.insert(tk.END, "DIAGNOSIS DETAILS\n\n", "header")
        
        crop_type = item["details"].get("crop_type", "N/A")
        disease_name = item["details"].get("disease_name", "N/A")
        confidence = item["details"].get("confidence", "N/A")

        icon = "⚠️" if "healthy" not in disease_name.lower() else "✅"
        self.details_text.insert(tk.END, f"Crop Type: {crop_type}\n")
        self.details_text.insert(tk.END, f"{icon} Disease: {disease_name}\n", "disease_icon")
        
        try:
            conf_val = float(confidence.replace("%", ""))
            if conf_val > 75:
                self.details_text.insert(tk.END, f"Confidence: {confidence}\n", "confidence_high")
            elif conf_val > 50:
                self.details_text.insert(tk.END, f"Confidence: {confidence}\n", "confidence_medium")
            else:
                self.details_text.insert(tk.END, f"Confidence: {confidence}\n", "confidence_low")
        except ValueError:
            self.details_text.insert(tk.END, f"Confidence: {confidence}\n")
        
        self.details_text.insert(tk.END, f"Date: {item['timestamp']}\n")
        if item.get("image"):
            self.details_text.insert(tk.END, f"Image File: {os.path.basename(item['image'])}\n")
        
        self.details_text.insert(tk.END, "\n" + "="*40 + "\n\n", "separator")

        # Display Disease Information
        self.details_text.insert(tk.END, "Disease Information:\n", "info_header")
        full_disease_key = f"{crop_type}_{disease_name.replace(' ', '_')}"
        info = self.controller.frames["DiagnosisPage"].disease_info.get(full_disease_key, self.controller.frames["DiagnosisPage"].disease_info.get("healthy"))
        
        if info:
            self.details_text.insert(tk.END, f"Description: {info['description']}\n\n", "info_content")
            self.details_text.insert(tk.END, f"Symptoms: {info['symptoms']}\n", "info_content")
        else:
            self.details_text.insert(tk.END, "No detailed information available for this disease.\n", "info_content")
        
        self.details_text.insert(tk.END, "\n" + "="*40 + "\n\n", "separator")

        # Display Treatment Suggestions
        self.details_text.insert(tk.END, "Treatment Suggestions:\n", "treatment_header")
        if item.get("treatment_suggestions"):
            for suggestion in item["treatment_suggestions"]:
                self.details_text.insert(tk.END, f"- {suggestion}\n", "treatment_bullet")
        else:
            self.details_text.insert(tk.END, "No specific treatment suggestions available.\n", "treatment_bullet")

        self.details_text.config(state=tk.DISABLED)
        self.show_history_image(item.get("image")) # Ensure image is displayed

    def show_history_image(self, image_path):
        if not image_path or not os.path.exists(image_path):
            self.image_preview.config(image="")
            self.image_preview.image = None
            return
            
        try:
            img = Image.open(image_path)
            img.thumbnail((300, 300), Image.Resampling.LANCZOS)
            img_tk = ImageTk.PhotoImage(img)
            
            self.image_preview.config(image=img_tk)
            self.image_preview.image = img_tk
        except Exception as e:
            logging.error(f"Error loading history image: {e}")
            self.image_preview.config(image="")
            self.image_preview.image = None
            self.controller._show_notification(f"Could not load image preview for {os.path.basename(image_path)}", "error")

    def view_full_image(self):
        if not self.selected_item_data or not self.selected_item_data.get("image") or not os.path.exists(self.selected_item_data["image"]):
            self.controller._show_notification("No valid image associated with selected record.", "warning")
            return
        
        try:
            webbrowser.open_new(self.selected_item_data["image"])
            self.controller._show_notification(f"Opening image: {os.path.basename(self.selected_item_data['image'])}", "info")
        except Exception as e:
            self.controller._show_notification(f"Failed to open image: {e}", "error")
            logging.error(f"Failed to open image {self.selected_item_data['image']}: {e}")

    def re_diagnose_image(self):
        if not self.selected_item_data or not self.selected_item_data.get("image") or not os.path.exists(self.selected_item_data["image"]):
            self.controller._show_notification("No valid image associated with selected record to re-diagnose.", "warning")
            return
        
        image_path = self.selected_item_data["image"]
        self.controller.show_frame("DiagnosisPage")
        # Pass the image path to the DiagnosisPage's upload_image method
        self.controller.frames["DiagnosisPage"].upload_image(image_path)
        self.controller._show_notification(f"Image sent to Diagnosis page: {os.path.basename(image_path)}", "info")


    def export_selected(self):
        if not self.selected_item_data:
            self.controller._show_notification("Please select an item to export.", "warning")
            return
            
        item = self.selected_item_data
        
        file_path = filedialog.asksaveasfilename(
            defaultextension=".pdf",
            filetypes=[("PDF files", "*.pdf"), ("Text files", "*.txt")],
            title="Save Diagnosis Report",
            initialfile=f"diagnosis_{item['timestamp'].replace(':', '-')}"
        )
        
        if file_path:
            try:
                if file_path.endswith(".pdf"):
                    self.export_to_pdf(item, file_path)
                else:
                    self.export_to_text(item, file_path)
                    
                self.controller._show_notification(f"Report saved to: {os.path.basename(file_path)}", "success")
            except Exception as e:
                self.controller._show_notification(f"Failed to export report: {e}", "error")

    def export_to_pdf(self, item, file_path):
        pdf = FPDF()
        pdf.set_font("Helvetica", "B", 16) # Changed font to Helvetica
        pdf.add_page()
        
        pdf.cell(0, 10, "Crop Disease Diagnosis Report", ln=True, align="C")
        pdf.ln(10)
        
        pdf.set_font("Helvetica", size=12) # Changed font to Helvetica
        pdf.cell(0, 10, f"Date: {item['timestamp']}", ln=True)
        if "image" in item:
            pdf.cell(0, 10, f"Image: {os.path.basename(item['image'])}", ln=True)
        pdf.ln(5)
        
        pdf.set_font("Helvetica", "B", 14) # Changed font to Helvetica
        pdf.cell(0, 10, "Diagnosis Results:", ln=True)
        pdf.set_font("Helvetica", size=12) # Changed font to Helvetica

        if "details" in item and item["details"]:
            crop_type = item["details"].get("crop_type", "N/A")
            disease_name = item["details"].get("disease_name", "N/A")
            confidence = item["details"].get("confidence", "N/A")
        else:
            diagnosis_lines = item["diagnosis"].split("\n")
            crop_type = next((line.split(":")[1].strip() for line in diagnosis_lines if "Crop Type:" in line), "N/A")
            disease_name = next((line.split(":")[1].strip() for line in diagnosis_lines if "Disease:" in line), "N/A")
            confidence = next((line.split(":")[1].strip() for line in diagnosis_lines if "Confidence:" in line), "N/A")
        
        pdf.cell(0, 7, f"Crop Type: {crop_type}", ln=True)
        pdf.cell(0, 7, f"Disease: {disease_name}", ln=True)
        pdf.cell(0, 7, f"Confidence: {confidence}", ln=True)
        pdf.ln(5)

        pdf.set_font("Helvetica", "B", 14) # Changed font to Helvetica
        pdf.cell(0, 10, "Treatment Suggestions:", ln=True)
        pdf.set_font("Helvetica", size=12) # Changed font to Helvetica

        if "treatment_suggestions" in item and item["treatment_suggestions"]:
            for suggestion in item["treatment_suggestions"]:
                pdf.multi_cell(0, 7, f"- {suggestion}")
        else:
            treatment_section_start = item["diagnosis"].find("Treatment Suggestions:")
            if treatment_section_start != -1:
                treatment_text = item["diagnosis"][treatment_section_start:].replace("Treatment Suggestions:", "").strip()
                for line in treatment_text.split("\n"):
                    if line.strip():
                        pdf.multi_cell(0, 7, line.replace("•", "-"))
            else:
                pdf.multi_cell(0, 7, "No specific treatment suggestions available for this diagnosis.")
        pdf.ln(10)
        
        if "image" in item and os.path.exists(item["image"]):
            temp_img_path = None
            try:
                img = Image.open(item["image"])
                img = img.convert("RGB")

                img_width, img_height = img.size
                max_pdf_width = 150
                max_pdf_height = 150

                if img_width > max_pdf_width or img_height > max_pdf_height:
                    img.thumbnail((max_pdf_width, max_pdf_height), Image.Resampling.LANCZOS)

                with tempfile.NamedTemporaryFile(delete=False, suffix=".jpg") as temp_file:
                    temp_img_path = temp_file.name
                    img.save(temp_img_path, format="JPEG")
                
                pdf.ln(5)
                pdf.cell(0, 10, "Image Preview:", ln=True)
                x_pos = (pdf.w - img.width) / 2
                pdf.image(temp_img_path, x=x_pos, w=img.width)
            except Exception as img_embed_e:
                logging.error(f"Failed to embed image in PDF for export: {img_embed_e}", exc_info=True)
                pdf.set_font("Helvetica", "I", 10) # Changed font to Helvetica
                pdf.cell(0, 10, "Error: Could not embed image in report.", ln=True, align="C")
            finally:
                if temp_img_path and os.path.exists(temp_img_path):
                    os.remove(temp_img_path)
        
        pdf.output(file_path)

    def export_to_text(self, item, file_path):
        with open(file_path, "w") as f:
            f.write("CROP DISEASE DIAGNOSIS REPORT\n")
            f.write("="*30 + "\n\n")
            f.write(f"Date: {item['timestamp']}\n")
            if "image" in item:
                f.write(f"Image: {os.path.basename(item['image'])}\n")
            f.write("\nDIAGNOSIS RESULTS:\n")
            f.write("="*30 + "\n")
            f.write(item["diagnosis"])

    def delete_selected(self):
        if not self.selected_item_data:
            self.controller._show_notification("Please select an item to delete.", "warning")
            return
            
        item_to_delete = self.selected_item_data
        
        if messagebox.askyesno("Confirm", "Delete this diagnosis record?"):
            if item_to_delete in self.controller.diagnosis_history:
                self.controller.diagnosis_history.remove(item_to_delete)
                self.refresh_history()
                self.clear_details_display() # Clear details after deletion
                self.controller._show_notification("Diagnosis record deleted.", "success")
            else:
                self.controller._show_notification("Selected record not found in history.", "error")


    def clear_history(self):
        if messagebox.askyesno("Confirm", "Are you sure you want to clear all diagnosis history? This action cannot be undone."):
            self.controller.diagnosis_history.clear()
            self.refresh_history()
            self.clear_details_display() # Clear details after clearing all history
            self.controller._show_notification("All diagnosis history records have been cleared.", "success")

    def export_all_history(self):
        if not self.controller.diagnosis_history:
            self.controller._show_notification("There is no history to export.", "info")
            return

        file_path = filedialog.asksaveasfilename(
            defaultextension=".json",
            filetypes=[("JSON files", "*.json"), ("All files", "*.*")],
            title="Export All Diagnosis History",
            initialfile=f"crop_diagnosis_history_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
        )

        if file_path:
            try:
                export_data = []
                for item in self.controller.diagnosis_history:
                    export_data.append({
                        "image": item.get("image"),
                        "diagnosis": item.get("diagnosis"),
                        "timestamp": item.get("timestamp"),
                        "details": item.get("details"),
                        "treatment_suggestions": item.get("treatment_suggestions")
                    })

                with open(file_path, 'w', encoding='utf-8') as f:
                    json.dump(export_data, f, indent=4)
                self.controller._show_notification(f"All history exported to: {os.path.basename(file_path)}", "success")
            except Exception as e:
                self.controller._show_notification(f"Failed to export history: {e}", "error")
                logging.error(f"History export error: {e}")

    def import_history(self):
        file_path = filedialog.askopenfilename(
            filetypes=[("JSON files", "*.json"), ("All files", "*.*")],
            title="Import Diagnosis History"
        )

        if file_path:
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    imported_data = json.load(f)
                
                if not isinstance(imported_data, list):
                    raise ValueError("Imported file does not contain a list of history records.")

                if self.controller.diagnosis_history:
                    if not messagebox.askyesno("Import History", "Do you want to append this history to your existing history? Click No to overwrite existing history."):
                        self.controller.diagnosis_history.clear()

                for item in imported_data:
                    if all(key in item for key in ["diagnosis", "timestamp"]):
                        if 'image' in item and not os.path.exists(item['image']):
                            item['image'] = None
                            logging.warning(f"Imported history item references missing image: {item.get('image_original_path', 'N/A')}")
                        self.controller.diagnosis_history.append(item)
                    else:
                        logging.warning(f"Skipping invalid history item during import: {item}")

                self.refresh_history()
                self.controller._show_notification(f"History imported from: {os.path.basename(file_path)}", "success")

            except json.JSONDecodeError as e:
                self.controller._show_notification(f"Invalid JSON file format: {e}", "error")
                logging.error(f"History import JSON error: {e}")
            except ValueError as e:
                self.controller._show_notification(str(e), "error")
                logging.error(f"History import data validation error: {e}")
            except Exception as e:
                self.controller._show_notification(f"Failed to import history: {e}", "error")
                logging.error(f"History import general error: {e}")


class FeedbackPage(ttk.Frame):
    def __init__(self, parent, controller):
        ttk.Frame.__init__(self, parent)
        self.controller = controller
        
        self.grid_rowconfigure(1, weight=1)
        self.grid_columnconfigure(0, weight=1)

        title_frame = ttk.Frame(self)
        title_frame.grid(row=0, column=0, columnspan=2, pady=(20, 10), padx=10, sticky="ew")

        title_label = ttk.Label(
            title_frame, 
            text="User Feedback & Support", 
            font=("Inter", 18, "bold")
        )
        title_label.pack(side=tk.LEFT)

        content_frame = ttk.Frame(self)
        content_frame.grid(row=1, column=0, sticky="nsew", padx=20, pady=10)
        content_frame.grid_columnconfigure(0, weight=1)

        type_frame = ttk.LabelFrame(content_frame, text="Feedback Type")
        type_frame.grid(row=0, column=0, sticky="ew", pady=(0, 15))
        
        self.feedback_type = tk.StringVar(value="suggestion")
        
        types = [
            ("Suggestion", "suggestion"),
            ("Bug Report", "bug"),
            ("Feature Request", "feature"),
            ("General Feedback", "general")
        ]
        
        for i, (text, value) in enumerate(types):
            rb = ttk.Radiobutton(
                type_frame,
                text=text,
                variable=self.feedback_type,
                value=value
            )
            rb.grid(row=0, column=i, padx=5, pady=5, sticky="w")

        contact_frame = ttk.Frame(content_frame)
        contact_frame.grid(row=1, column=0, sticky="ew", pady=(0, 15))
        
        ttk.Label(contact_frame, text="Contact Email (optional):").pack(side=tk.LEFT)
        self.email_entry = ttk.Entry(contact_frame, width=30)
        self.email_entry.pack(side=tk.LEFT, padx=5)

        self.rating_frame = ttk.LabelFrame(content_frame, text="Rating")
        self.rating_frame.grid(row=2, column=0, sticky="ew", pady=(0, 15))
        
        self.rating = tk.IntVar(value=3)
        rating_labels = ["Poor", "Fair", "Good", "Very Good", "Excellent"]
        
        for i in range(5):
            rb = ttk.Radiobutton(
                self.rating_frame,
                text=str(i+1),
                variable=self.rating,
                value=i+1
            )
            rb.grid(row=0, column=i*2, padx=2)
            
            if i < len(rating_labels):
                ttk.Label(
                    self.rating_frame,
                    text=rating_labels[i],
                    font=("Inter", 8)
                ).grid(row=0, column=i*2+1, padx=2, sticky="w")

        text_frame = ttk.LabelFrame(content_frame, text="Your Feedback")
        text_frame.grid(row=3, column=0, sticky="nsew", pady=(0, 15))
        text_frame.grid_rowconfigure(0, weight=1)
        text_frame.grid_columnconfigure(0, weight=1)
        
        self.feedback_text = tk.Text(
            text_frame, 
            wrap="word",
            font=("Inter", 11),
            padx=5,
            pady=5,
            height=10
        )
        
        scrollbar = ttk.Scrollbar(text_frame, command=self.feedback_text.yview)
        scrollbar.grid(row=0, column=1, sticky="ns")
        self.feedback_text.config(yscrollcommand=scrollbar.set)
        self.feedback_text.grid(row=0, column=0, sticky="nsew")
        
        self.feedback_text.insert("1.0", "Please describe your feedback in detail...")
        self.feedback_text.config(fg="grey")
        
        def on_focus_in(event):
            if self.feedback_text.get("1.0", "end-1c") == "Please describe your feedback in detail...":
                self.feedback_text.delete("1.0", "end")
                self.feedback_text.config(fg="black")
        
        def on_focus_out(event):
            if not self.feedback_text.get("1.0", "end-1c").strip():
                self.feedback_text.insert("1.0", "Please describe your feedback in detail...")
                self.feedback_text.config(fg="grey")
        
        self.feedback_text.bind("<FocusIn>", on_focus_in)
        self.feedback_text.bind("<FocusOut>", on_focus_out)

        attach_frame = ttk.Frame(content_frame)
        attach_frame.grid(row=4, column=0, sticky="ew", pady=(0, 15))
        
        self.attach_screenshot = tk.BooleanVar(value=False)
        ttk.Checkbutton(
            attach_frame,
            text="Attach screenshot (if applicable)",
            variable=self.attach_screenshot,
            command=self.toggle_screenshot_option
        ).pack(side=tk.LEFT)
        
        self.screenshot_path = None
        self.screenshot_button = ttk.Button(
            attach_frame,
            text="Select Screenshot",
            command=self.attach_screenshot_file,
            state=tk.DISABLED
        )
        self.screenshot_button.pack(side=tk.LEFT, padx=5)

        self.screenshot_filename_label = ttk.Label(attach_frame, text="", font=("Inter", 9, "italic"))
        self.screenshot_filename_label.pack(side=tk.LEFT, padx=5)

        button_frame = ttk.Frame(content_frame)
        button_frame.grid(row=5, column=0, sticky="ew", pady=(10, 0))
        
        self.submit_button = ttk.Button(
            button_frame,
            text="Submit Feedback",
            command=self.submit_feedback,
            style='TButton'
        )
        self.submit_button.pack(side=tk.LEFT, padx=5, expand=True, fill=tk.X)
        
        self.reset_button = ttk.Button(
            button_frame,
            text="Reset Form",
            command=self.reset_form,
            style='TButton'
        )
        self.reset_button.pack(side=tk.LEFT, padx=5, expand=True, fill=tk.X)

        back_button = ttk.Button(
            button_frame,
            text="Back",
            command=lambda: controller.show_frame("HomePage"),
            style='TButton'
        )
        back_button.pack(side=tk.LEFT, padx=5, expand=True, fill=tk.X)

        self.feedback_type.trace_add("write", self.update_feedback_ui)

    def toggle_screenshot_option(self):
        if self.attach_screenshot.get():
            self.screenshot_button.config(state=tk.NORMAL)
        else:
            self.screenshot_button.config(state=tk.DISABLED)
            self.screenshot_path = None
            self.screenshot_filename_label.config(text="") # Clear filename when unchecked

    def attach_screenshot_file(self):
        file_types = [
            ("Image files", "*.jpg;*.jpeg;*.png;*.bmp"),
            ("All files", "*.*")
        ]
        file_path = filedialog.askopenfilename(
            title="Select Screenshot",
            filetypes=file_types
        )
        
        if file_path:
            self.screenshot_path = file_path
            self.screenshot_filename_label.config(text=os.path.basename(file_path))
            self.controller._show_notification(f"Screenshot attached: {os.path.basename(file_path)}", "info")
        else:
            self.screenshot_filename_label.config(text="") # Clear if selection is cancelled
            self.screenshot_path = None
            self.controller._show_notification("Screenshot attachment cancelled.", "info")


    def update_feedback_ui(self, *args):
        feedback_type = self.feedback_type.get()
        
        if feedback_type in ["suggestion", "general"]:
            self.rating_frame.grid()
        else:
            self.rating_frame.grid_remove()
        
        placeholders = {
            "suggestion": "Please describe your suggestion for improvement...",
            "bug": "Please describe the bug you encountered, including steps to reproduce...",
            "feature": "Please describe the feature you'd like to see added...",
            "general": "Please share your general feedback about the application..."
        }
        
        current_text = self.feedback_text.get("1.0", "end-1c")
        if current_text.startswith("Please describe") or not current_text.strip():
            self.feedback_text.delete("1.0", "end")
            self.feedback_text.insert("1.0", placeholders[feedback_type])
            self.feedback_text.config(fg="grey")

    def reset_form(self):
        self.feedback_type.set("suggestion")
        self.email_entry.delete(0, tk.END)
        self.rating.set(3)
        self.feedback_text.delete("1.0", tk.END)
        self.feedback_text.insert("1.0", "Please describe your feedback in detail...")
        self.feedback_text.config(fg="grey")
        self.attach_screenshot.set(False)
        self.screenshot_path = None
        self.screenshot_button.config(state=tk.DISABLED)
        self.screenshot_filename_label.config(text="")
        self.controller._show_notification("Feedback form reset.", "info")


    def submit_feedback(self):
        feedback = self.feedback_text.get("1.0", tk.END).strip()
        
        if feedback == "" or feedback.startswith("Please describe"):
            self.controller._show_notification("Please enter your feedback before submitting.", "warning")
            return

        email = self.email_entry.get().strip()
        if email and not re.match(r"[^@]+@[^@]+\.[^@]+", email):
            self.controller._show_notification("Please enter a valid email address.", "warning")
            return

        if not messagebox.askyesno("Confirm Submission", "Are you sure you want to submit this feedback?"):
            return # User cancelled

        feedback_dir = "feedback"
        os.makedirs(feedback_dir, exist_ok=True)
        
        feedback_data = {
            "type": self.feedback_type.get(),
            "text": feedback,
            "email": email,
            "rating": self.rating.get() if self.feedback_type.get() in ["suggestion", "general"] else None,
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "version": "1.0",
            "platform": sys.platform
        }
        
        if self.attach_screenshot.get() and self.screenshot_path and os.path.exists(self.screenshot_path):
            try:
                screenshots_dir = os.path.join(feedback_dir, "screenshots")
                os.makedirs(screenshots_dir, exist_ok=True)
                
                screenshot_filename = f"screenshot_{datetime.now().strftime('%Y%m%d_%H%M%S')}_{os.path.basename(self.screenshot_path)}"
                screenshot_path = os.path.join(screenshots_dir, screenshot_filename)
                shutil.copy2(self.screenshot_path, screenshot_path)
                
                feedback_data["screenshot"] = screenshot_filename
            except Exception as e:
                logging.error(f"Failed to save screenshot: {e}")
                self.controller._show_notification(f"Failed to save screenshot: {e}", "error")
                # Continue submission without screenshot if saving fails

        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        file_path = os.path.join(feedback_dir, f"feedback_{timestamp}.json")
        
        try:
            self.submit_button.config(state=tk.DISABLED, text="Submitting...")
            self.update() # Force UI update
            
            # Simulate network delay or actual server call here
            time.sleep(1) 
            
            with open(file_path, "w", encoding='utf-8') as f:
                json.dump(feedback_data, f, indent=2)
            
            self.controller._show_notification("Feedback submitted successfully. Thank you!", "success")
            self.reset_form() # Reset the form after successful submission
            
        except Exception as e:
            self.controller._show_notification(f"Failed to submit feedback: {e}", "error")
            logging.error(f"Feedback submission error: {e}")
        finally:
            self.submit_button.config(state=tk.NORMAL, text="Submit Feedback")

    def send_feedback_to_server(self, feedback_data):
        # This function is a placeholder and would require a real API endpoint
        try:
            import requests
            
            response = requests.post(
                "https://your-api-endpoint.com/feedback", # Replace with your actual API endpoint
                json=feedback_data,
                timeout=5
            )
            
            if response.status_code != 200:
                raise Exception(f"Server returned status {response.status_code}")
                
        except Exception as e:
            logging.error(f"Failed to send feedback to server: {e}")
            raise Exception("Couldn't connect to server. Feedback saved locally.")


class HelpPage(ttk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent)
        self.controller = controller

        self.grid_rowconfigure(0, weight=1)
        self.grid_columnconfigure(0, weight=1)

        self.canvas = tk.Canvas(self, highlightthickness=0, bg=self.controller.colors['bg'])
        self.canvas.is_help_canvas = True
        self.scrollbar = ttk.Scrollbar(self, orient="vertical", command=self.canvas.yview)
        self.canvas.configure(yscrollcommand=self.scrollbar.set)

        self.scrollable_frame = ttk.Frame(self.canvas)
        self.canvas_frame = self.canvas.create_window((0, 0), window=self.scrollable_frame, anchor="nw")

        self.bind("<Configure>", self._resize_canvas)

        self.canvas.grid(row=0, column=0, sticky="nsew")
        self.scrollbar.grid(row=0, column=1, sticky="ns")

        self.scrollable_frame.grid_columnconfigure(0, weight=1)

        self.canvas.bind("<MouseWheel>", self._on_mousewheel)
        self.scrollable_frame.bind("<MouseWheel>", self._on_mousewheel)
        self.canvas.bind("<Button-4>", self._on_mousewheel)
        self.canvas.bind("<Button-5>", self._on_mousewheel)
        self.scrollable_frame.bind("<Button-4>", self._on_mousewheel)
        self.scrollable_frame.bind("<Button-5>", self._on_mousewheel)

        self._build_help_content()

        self.scrollable_frame.update_idletasks()
        self.canvas.config(scrollregion=(0, 0, self.scrollable_frame.winfo_width(), self.scrollable_frame.winfo_reqheight()))


    def _resize_canvas(self, event):
        canvas_width = event.width
        self.canvas.itemconfig(self.canvas_frame, width=canvas_width)
        self.scrollable_frame.update_idletasks()
        self.canvas.config(scrollregion=(0, 0, self.scrollable_frame.winfo_width(), self.scrollable_frame.winfo_reqheight()))

    def _on_mousewheel(self, event):
        if sys.platform == "darwin":
            self.canvas.yview_scroll(int(-1*(event.delta)), "units")
        elif sys.platform == "win32":
            self.canvas.yview_scroll(int(-1*(event.delta/120)), "units")
        else:
            if event.num == 4:
                self.canvas.yview_scroll(-1, "units")
            elif event.num == 5:
                self.canvas.yview_scroll(1, "units")

    def _build_help_content(self):
        header_frame = ttk.Frame(self.scrollable_frame)
        header_frame.pack(fill=tk.X, padx=20, pady=(20, 10))
        
        try:
            logo_path = self.controller.get_resource_path(os.path.join("assets", r"C:\Users\ayush\OneDrive\Desktop\project 1\CropDiseaseApp\assets\icons8-chaff-48.png"))

            if os.path.exists(logo_path):
                logo_img = Image.open(logo_path).resize((64, 64), Image.Resampling.LANCZOS)
                self.logo_photo = ImageTk.PhotoImage(logo_img)
                ttk.Label(header_frame, image=self.logo_photo).pack(side=tk.LEFT, padx=(0, 15))
        except Exception as e:
            logging.warning(f"Logo load failed in help page: {e}")

        title_frame = ttk.Frame(header_frame)
        title_frame.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
        
        title_label_main = ttk.Label(
            title_frame, 
            text="Crop Disease Diagnosis Guide", 
            font=("Inter", 22, "bold"),
            foreground=self.controller.colors['help_page_title_fg']
        )
        title_label_main.is_help_title = True
        title_label_main.pack(anchor="w")
        
        subtitle_label_main = ttk.Label(
            title_frame, 
            text="Comprehensive user manual and troubleshooting guide", 
            font=("Inter", 12),
            foreground=self.controller.colors['help_page_text_fg']
        )
        subtitle_label_main.is_help_subtitle = True
        subtitle_label_main.pack(anchor="w")

        search_frame = ttk.Frame(self.scrollable_frame)
        search_frame.pack(fill=tk.X, padx=20, pady=(0, 20))
        
        self.search_var = tk.StringVar()
        search_entry = ttk.Entry(
            search_frame,
            textvariable=self.search_var,
            font=("Inter", 12)
        )
        search_entry.pack(side=tk.LEFT, fill=tk.X, expand=True, padx=(0, 5))
        
        ttk.Button(
            search_frame,
            text="Search",
            command=self._search_help,
            style='TButton'
        ).pack(side=tk.LEFT)

        # Table of Contents
        toc_frame = ttk.LabelFrame(self.scrollable_frame, text="Table of Contents")
        toc_frame.pack(fill=tk.X, padx=20, pady=(0, 15))
        
        self.toc_links = [] # To store references to the TOC labels
        
        self.help_sections = [
            {
                "title": "Getting Started",
                "content": [
                    ("Welcome to Crop Disease Diagnosis", "This AI-powered application helps farmers and agricultural professionals identify diseases in crop leaves through image analysis."),
                    ("System Requirements", "• Python 3.8 or higher\n• 4GB RAM minimum\n• Webcam or camera for live capture\n• Supported image formats: JPG, PNG, BMP"),
                    ("First Time Setup", "1. Install all required dependencies\n2. Place model files in the 'models' folder\n3. Configure your camera if using live capture")
                ],
                "expanded": True,
                "frame": None # Will be set later
            },
            {
                "title": "Image Capture Guidelines",
                "content": [
                    ("Optimal Image Quality", "• Capture in well-lit conditions\n• Focus on the leaf surface\n• Include both healthy and affected areas\n• Avoid shadows and glare"),
                    ("Recommended Practices", "- Place leaf on neutral background\n- Include scale reference if possible\n- Capture multiple angles for severe cases"),
                    ("Common Mistakes", "× Blurry or out-of-focus images\n× Images with multiple leaves\n× Extreme close-ups without context\n× Poor lighting conditions")
                ],
                "expanded": True,
                "frame": None
            },
            {
                "title": "Diagnosis Process",
                "content": [
                    ("Step-by-Step Guide", "1. Navigate to Diagnosis page\n2. Upload or capture image\n3. Adjust image if needed\n4. View results\n5. Save or export report"),
                    ("Understanding Results", "• Confidence scores indicate prediction reliability\n• Multiple results may appear for similar diseases\n• Low confidence suggests retaking the image"),
                    ("Troubleshooting", "If you get poor results:\n- Try different image adjustments\n- Capture a clearer image\n- Check for system updates")
                ],
                "expanded": False,
                "frame": None
            },
            {
                "title": "Advanced Features",
                "content": [
                    ("Image Enhancement Tools", "• Brightness/Contrast adjustment\n• Color saturation controls\n• Sharpening filters\n• Side-by-side comparison"),
                    ("Batch Processing", "- Select multiple images at once\n- View summary reports\n- Export all results to CSV"),
                    ("History Management", "• View past diagnoses\n• Filter by date/crop type\n•• Export complete history")
                ],
                "expanded": False,
                "frame": None
            },
            {
                "title": "Troubleshooting",
                "content": [
                    ("Common Issues", "• Model not loading: Check file paths\n• Slow performance: Close other applications\n• Incorrect results: Verify image quality"),
                    ("Error Messages", "- 'Invalid image': Try different format\n- 'Low confidence': Retake image\n- 'Model error': Restart application"),
                    ("FAQ", "Q: Can I use this for any crop?\nA: Currently supports 15 major crops (see full list)\n\nQ: How accurate are the results?\nA: Average 92% accuracy in lab conditions")
                ],
                "expanded": False,
                "frame": None
            },
            {
                "title": "Support & Resources",
                "content": [
                    ("Contact Information", "• Email: support@cropdiagnosis.com\n• Phone: +1 (800) 555-AGRO\n• Online chat: Available 9AM-5PM EST"),
                    ("Additional Resources", "- Video tutorials on YouTube\n- PDF user manual download\n- Community forums"),
                    ("Feedback", "We value your input!\nPlease report any issues or suggestions\nthrough the Feedback page")
                ],
                "expanded": False,
                "frame": None
            }
        ]

        for i, section in enumerate(self.help_sections):
            # Create TOC link
            toc_link = ttk.Label(
                toc_frame,
                text=f"{i+1}. {section['title']}",
                font=("Inter", 11, "underline"),
                foreground=self.controller.colors['fg'],
                cursor="hand2"
            )
            toc_link.is_toc_link = True # Custom attribute for color updates
            toc_link.pack(anchor="w", padx=10, pady=2)
            toc_link.bind("<Button-1>", lambda e, s=section: self._scroll_to_section(s))
            self.toc_links.append(toc_link) # Store reference

            # Create the actual section content
            self._create_section(section)

        button_frame = ttk.Frame(self.scrollable_frame)
        button_frame.pack(fill=tk.X, padx=20, pady=(20, 30))
        
        ttk.Button(
            button_frame,
            text="← Back to Home",
            command=lambda: self.controller.show_frame("HomePage"),
            style='TButton'
        ).pack(side=tk.LEFT, ipady=5, ipadx=20)

        ttk.Button(
            button_frame,
            text="Print This Guide",
            command=self._print_guide,
            style='TButton'
        ).pack(side=tk.RIGHT, ipady=5, ipadx=20)

    def _create_section(self, section):
        section_frame = ttk.Frame(
            self.scrollable_frame,
            relief="solid",
            borderwidth=1,
            style='Card.TFrame'
        )
        section_frame.pack(fill=tk.X, padx=20, pady=5)

        toggle_header_frame = ttk.Frame(section_frame)
        toggle_header_frame.pack(fill=tk.X, padx=10, pady=5)
        
        toggle_btn = ttk.Button(
            toggle_header_frame,
            text="+" if not section["expanded"] else "-",
            width=3,
            command=lambda s=section, f=section_frame: self._toggle_section(s, f),
            style='Toggle.TButton'
        )
        toggle_btn.pack(side=tk.LEFT, padx=(0, 10))

        title_label = ttk.Label(
            toggle_header_frame,
            text=section["title"],
            font=("Inter", 14, "bold"),
            foreground=self.controller.colors['help_page_text_fg'],
            anchor="w"
        )
        title_label.is_help_section_title = True
        title_label.pack(side=tk.LEFT, fill=tk.X, expand=True)

        content_area = ttk.Frame(section_frame)
        if section["expanded"]:
            content_area.pack(fill=tk.X, padx=20, pady=(0, 10))
        
        for item in section["content"]:
            if isinstance(item, tuple):
                subtitle, content = item
                subtitle_label = ttk.Label(
                    content_area,
                    text=subtitle,
                    font=("Inter", 12, "bold"),
                    foreground=self.controller.colors['help_page_text_fg'],
                    anchor="w"
                )
                subtitle_label.is_help_content = True
                subtitle_label.pack(fill=tk.X, pady=(10, 2))
                
                content_label = ttk.Label(
                    content_area,
                    text=content,
                    font=("Inter", 11),
                    wraplength=700,
                    foreground=self.controller.colors['help_page_text_fg'],
                    anchor="w",
                    justify="left"
                )
                content_label.is_help_content = True
                content_label.pack(fill=tk.X, pady=(0, 10))
            else:
                simple_content_label = ttk.Label(
                    content_area,
                    text=item,
                    font=("Inter", 11),
                    wraplength=700,
                    foreground=self.controller.colors['help_page_text_fg'],
                    anchor="w",
                    justify="left"
                )
                simple_content_label.is_help_content = True
                simple_content_label.pack(fill=tk.X, pady=2)

        section["frame"] = section_frame
        section["content_area"] = content_area
        section["toggle_btn"] = toggle_btn

    def _toggle_section(self, section, frame):
        section["expanded"] = not section["expanded"]
        
        if section["expanded"]:
            section["content_area"].pack(fill=tk.X, padx=20, pady=(0, 10))
            section["toggle_btn"].configure(text="-")
        else:
            section["content_area"].pack_forget()
            section["toggle_btn"].configure(text="+")
        
        self.scrollable_frame.update_idletasks()
        self.canvas.configure(scrollregion=(0, 0, self.scrollable_frame.winfo_width(), self.scrollable_frame.winfo_reqheight()))

    def _scroll_to_section(self, section):
        # Ensure the section is expanded before scrolling
        if not section["expanded"]:
            self._toggle_section(section, section["frame"])
        
        # Update idle tasks to ensure correct geometry information
        self.scrollable_frame.update_idletasks()
        
        # Calculate the relative position of the section frame within the scrollable frame
        section_y_pos = section["frame"].winfo_y()
        
        # Calculate the fraction to scroll to
        scroll_fraction = section_y_pos / self.scrollable_frame.winfo_height()
        
        # Clamp the fraction between 0 and 1
        scroll_fraction = max(0.0, min(1.0, scroll_fraction))
        
        self.canvas.yview_moveto(scroll_fraction)
        self.controller._show_notification(f"Scrolled to: {section['title']}", "info")


    def _search_help(self):
        query = self.search_var.get().lower()
        if not query:
            return
            
        found_match = False
        for section in self.help_sections:
            content_text = "\n".join(
                f"{item[0]}\n{item[1]}" if isinstance(item, tuple) else item 
                for item in section["content"]
            ).lower()
            
            if query in content_text or query in section["title"].lower():
                if not section["expanded"]:
                    self._toggle_section(section, section["frame"])
                
                # Scroll to the found section
                self._scroll_to_section(section)
                found_match = True
                break
        
        if not found_match:
            self.controller._show_notification(f"No results found for '{query}'", "warning")

    
    def _print_guide(self):
        if not messagebox.askyesno("Confirm Print", "Do you want to generate a printable PDF of this guide?"):
            return

        temp_html_file = None
        temp_pdf_path = None
        try:
            html_template = """<!DOCTYPE html>
    <html>
    <head>
        <meta charset="UTF-8">
        <title>Crop Disease Diagnosis Guide</title>
        <style>
            body {
                line-height: 1.6;
                margin: 20px;
                color: #1B5E20;
            }
            h1 {
                color: #2E7D32;
                text-align: center;
            }
            h2 {
                color: #1B5E20;
                border-bottom: 1px solid #ddd;
                padding-bottom: 5px;
            }
            .section {
                margin-bottom: 20px;
            }
            .subtitle {
                font-weight: bold;
                color: #1B5E20;
            }
            .content {
                margin-left: 15px;
                margin-bottom: 10px;
                white-space: pre-line;
            }
        </style>
    </head>
    <body>
        <h1>Crop Disease Diagnosis Guide</h1>
        <p>Generated on {date}</p>
        {content}
    </body>
    </html>"""

            content_sections = []
            for section in self.help_sections:
                section_html = f"<div class='section'><h2>{section['title']}</h2>"
            
                for item in section["content"]:
                    if isinstance(item, tuple):
                        subtitle, content = item
                        # Escape HTML characters in content
                        content = content.replace('&', '&amp;').replace('<', '&lt;').replace('>', '&gt;')
                        section_html += f"<div class='subtitle'>{subtitle}</div><div class='content'>{content}</div>"
                    else:
                        # Escape HTML characters in simple content
                        item = item.replace('&', '&amp;').replace('<', '&lt;').replace('>', '&gt;')
                        section_html += f"<div class='content'>{item}</div>"
            
                section_html += "</div>"
                content_sections.append(section_html)
        
            html_content = html_template.format(
                date=datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                content="\n".join(content_sections))
        
            # Use FPDF to render HTML to PDF
            pdf = FPDF()
            pdf.add_page()
            pdf.set_font("Helvetica", size=12) # Set a default font for FPDF rendering
            
            # Write the HTML content to a temporary file
            temp_html_file = os.path.join(tempfile.gettempdir(), "temp_help_guide.html")
            with open(temp_html_file, "w", encoding='utf-8') as f:
                f.write(html_content)

            # FPDF's write_html method can parse basic HTML
            # Note: FPDF's HTML support is limited; complex CSS/JS won't work.
            # For more robust HTML to PDF, consider external libraries like pdfkit (which uses wkhtmltopdf)
            # but that adds external dependencies not available in this environment.
            # The current approach relies on FPDF's built-in HTML parser.
            pdf.write_html(html_content)

            temp_pdf_path = os.path.join(tempfile.gettempdir(), "crop_help_guide.pdf")
            pdf.output(temp_pdf_path)
        
            import webbrowser
            webbrowser.open(f"file://{temp_pdf_path}")
        
            self.controller._show_notification("The guide has been generated and opened in your browser.", "success", duration=5000)
        
        except Exception as e:
            self.controller._show_notification(f"Could not generate printable guide: {str(e)}", "error", duration=5000)
            logging.error(f"Print guide error: {e}")
        finally:
            # Clean up temporary HTML file
            if temp_html_file and os.path.exists(temp_html_file):
                os.remove(temp_html_file)
            # Give a moment for the external PDF viewer to open before deleting the PDF
            if temp_pdf_path and os.path.exists(temp_pdf_path):
                threading.Timer(2, lambda: os.remove(temp_pdf_path)).start()


class AboutPage(ttk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent)
        self.controller = controller
        
        # Configure scrollable frame for content that might exceed window height
        self.canvas = tk.Canvas(self, borderwidth=0, highlightthickness=0, bg=self.controller.colors['about_page_bg'])
        self.canvas.is_about_canvas = True
        self.scrollbar = ttk.Scrollbar(self, orient="vertical", command=self.canvas.yview)
        
        self.scrollable_frame = ttk.Frame(self.canvas, style='AboutPage.TFrame') 
        self.canvas_frame_id = self.canvas.create_window((0, 0), window=self.scrollable_frame, anchor="nw")
        self.canvas.configure(yscrollcommand=self.scrollbar.set)

        self.scrollable_frame.bind(
            "<Configure>",
            lambda e: self.canvas.configure(
                scrollregion=self.canvas.bbox("all")
            )
        )
        
        self.canvas.pack(side="left", fill="both", expand=True)
        self.scrollbar.pack(side="right", fill="y")

        self.bind("<Configure>", self._on_frame_configure)
        self._build_about_content()

        self.scrollable_frame.update_idletasks()
        self.canvas.config(scrollregion=(0, 0, self.scrollable_frame.winfo_width(), self.scrollable_frame.winfo_reqheight()))
        
    def _on_frame_configure(self, event):
        self.canvas.itemconfig(self.canvas_frame_id, width=event.width)
        self.scrollable_frame.update_idletasks()
        self.canvas.config(scrollregion=(0, 0, self.scrollable_frame.winfo_width(), self.scrollable_frame.winfo_reqheight()))

    def _on_mousewheel(self, event):
        if sys.platform == "darwin":
            self.canvas.yview_scroll(int(-1*(event.delta)), "units")
        elif sys.platform == "win32":
            self.canvas.yview_scroll(int(-1*(event.delta/120)), "units")
        else:
            if event.num == 4:
                self.canvas.yview_scroll(-1, "units")
            elif event.num == 5:
                self.canvas.yview_scroll(1, "units")

    def _build_about_content(self):
        # Header Frame with Logo and Title
        header_frame = ttk.Frame(self.scrollable_frame, style='AboutPage.TFrame')
        header_frame.pack(fill=tk.X, padx=20, pady=(20, 10))

        # Logo - Fixed version with proper reference maintenance
        try:
            logo_path = self.controller.get_resource_path(os.path.join("assets", r"C:\Users\ayush\OneDrive\Desktop\project 1\CropDiseaseApp\assets\icons8-chaff-48.png"))
            if os.path.exists(logo_path):
                logo_img = Image.open(logo_path).resize((64, 64), Image.Resampling.LANCZOS)
                self.logo_photo = ImageTk.PhotoImage(logo_img)  # Maintain reference
                logo_label = ttk.Label(
                    header_frame, 
                    image=self.logo_photo,
                    style='AboutPage.TFrame'
                )
                logo_label.pack(side=tk.LEFT, padx=(0, 15))
            else:
                # Fallback if image not found
                ttk.Label(
                    header_frame,
                    text="[Logo]",
                    font=("Inter", 10),
                    style='AboutPage.TFrame'
                ).pack(side=tk.LEFT, padx=(0, 15))
                logging.warning(f"Logo image not found at: {logo_path}")
        except Exception as e:
            logging.error(f"Error loading logo: {e}")
            # Fallback if error occurs
            ttk.Label(
                header_frame,
                text="[Logo Error]",
                font=("Inter", 10),
                foreground="red",
                style='AboutPage.TFrame'
            ).pack(side=tk.LEFT, padx=(0, 15))

        # Main Title
        self.about_title_label = ttk.Label(
            header_frame,
            text="About Crop Disease Detection App",
            font=("Inter", 22, "bold"),
            foreground=self.controller.colors['about_page_title_fg'],
            style='AboutPageTitle.TLabel'
        )
        self.about_title_label.pack(fill=tk.X, expand=True, anchor="w")

        # Application Overview Section
        ttk.Label(
            self.scrollable_frame,
            text="Application Overview",
            font=("Inter", 13, "bold"),
            foreground=self.controller.colors['about_page_title_fg'],
            style='AboutPageSectionTitle.TLabel'
        ).pack(fill=tk.X, padx=20, pady=(10, 0), anchor="w")
        
        ttk.Label(
            self.scrollable_frame,
            text=(
                "The Crop Disease Detection App is an innovative tool designed to assist farmers "
                "and agricultural enthusiasts in identifying common plant leaf diseases using "
                "advanced machine learning models. Our goal is to provide quick, accessible, "
                "and preliminary diagnoses to help in timely intervention and better crop management."
            ),
            wraplength=700,
            justify="left",
            foreground=self.controller.colors['about_page_text_fg'],
            style='AboutPageContent.TLabel'
        ).pack(fill=tk.X, padx=30, pady=(1, 10), anchor="w")

        # Key Features Section
        ttk.Label(
            self.scrollable_frame,
            text="Key Features:",
            font=("Inter", 13, "bold"),
            foreground=self.controller.colors['about_page_title_fg'],
            style='AboutPageSectionTitle.TLabel'
        ).pack(fill=tk.X, padx=20, pady=(10, 0), anchor="w")

        features = [
            "• AI-powered disease detection for various crops.",
            "• Image upload and live camera capture capabilities.",
            "• Image enhancement tools for better analysis.",
            "• Detailed diagnosis reports with confidence scores and treatment suggestions.",
            "• History tracking of past diagnoses.",
            "• User feedback system for continuous improvement."
        ]
        for feature in features:
            ttk.Label(
                self.scrollable_frame,
                text=feature,
                font=("Inter", 11),
                wraplength=700,
                justify="left",
                foreground=self.controller.colors['about_page_text_fg'],
                style='AboutPageContent.TLabel'
            ).pack(fill=tk.X, padx=30, pady=1, anchor="w")

        # Development Information Section
        ttk.Label(
            self.scrollable_frame,
            text="Development Information:",
            font=("Inter", 13, "bold"),
            foreground=self.controller.colors['about_page_title_fg'],
            style='AboutPageSectionTitle.TLabel'
        ).pack(fill=tk.X, padx=20, pady=(10, 0), anchor="w")

        ttk.Label(
            self.scrollable_frame,
            text=f"Developed by: [AYUSH CHOUBEY]\nVersion: 1.0.0\nRelease Date: {datetime.now().strftime('%B %d, %Y')}",
            font=("Inter", 11),
            wraplength=700,
            justify="left",
            foreground=self.controller.colors['about_page_text_fg'],
            style='AboutPageContent.TLabel'
        ).pack(fill=tk.X, padx=30, pady=1, anchor="w")

        # Technologies Used Section
        ttk.Label(
            self.scrollable_frame,
            text="Technologies Used:",
            font=("Inter", 13, "bold"),
            foreground=self.controller.colors['about_page_title_fg'],
            style='AboutPageSectionTitle.TLabel'
        ).pack(fill=tk.X, padx=20, pady=(10, 0), anchor="w")
        
        technologies = [
            "• Python (Programming Language)",
            "• Tkinter (GUI Framework)",
            "• TensorFlow/Keras (Machine Learning)",
            "• Pillow (PIL Fork - Image Processing)",
            "• OpenCV (Computer Vision)",
            "• NumPy (Numerical Computing)",
            "• FPDF (PDF Generation)"
        ]
        for tech in technologies:
            ttk.Label(
                self.scrollable_frame,
                text=tech,
                font=("Inter", 11),
                wraplength=700,
                justify="left",
                foreground=self.controller.colors['about_page_text_fg'],
                style='AboutPageContent.TLabel'
            ).pack(fill=tk.X, padx=30, pady=1, anchor="w")

        # Contact Us Section
        ttk.Label(
            self.scrollable_frame,
            text="Contact Us:",
            font=("Inter", 13, "bold"),
            foreground=self.controller.colors['about_page_title_fg'],
            style='AboutPageSectionTitle.TLabel'
        ).pack(fill=tk.X, padx=20, pady=(10, 0), anchor="w")
        
        ttk.Label(
            self.scrollable_frame,
            text="For support or inquiries, please contact:",
            wraplength=700,
            justify="left",
            foreground=self.controller.colors['about_page_text_fg'],
            style='AboutPageContent.TLabel'
        ).pack(fill=tk.X, padx=30, pady=(1, 5), anchor="w")
        
        # Hyperlink for email
        email_label = ttk.Label(
            self.scrollable_frame,
            text="ayushchoubey800@gmail.com",
            wraplength=700,
            justify="left",
            foreground="blue",
            cursor="hand2",
            style='AboutPageContent.TLabel'
        )
        email_label.pack(fill=tk.X, padx=30, pady=(1, 5), anchor="w")
        email_label.bind("<Button-1>", lambda e: webbrowser.open_new_tab("mailto:ayushchoubey800@gmail.com"))

        # Hyperlink for website
        website_label = ttk.Label(
            self.scrollable_frame,
            text="Visit our website/repository",
            wraplength=700,
            justify="left",
            foreground="blue",
            cursor="hand2",
            style='AboutPageContent.TLabel'
        )
        website_label.pack(fill=tk.X, padx=30, pady=(1, 10), anchor="w")
        website_label.bind("<Button-1>", lambda e: webbrowser.open_new_tab("https://www.linkedin.com/in/ayush-choubey-b2a366281?utm_source=share&utm_campaign=share_via&utm_content=profile&utm_medium=android_app"))

        # Disclaimer Section
        ttk.Label(
            self.scrollable_frame,
            text="Disclaimer:",
            font=("Inter", 13, "bold"),
            foreground=self.controller.colors['about_page_title_fg'],
            style='AboutPageSectionTitle.TLabel'
        ).pack(fill=tk.X, padx=20, pady=(10, 0), anchor="w")

        ttk.Label(
            self.scrollable_frame,
            text=(
                "This application is a diagnostic aid and should not replace professional agricultural advice. "
                "Always consult with a qualified agronomist or plant pathologist for definitive diagnoses and treatment plans."
            ),
            font=("Inter", 11, "italic"),
            wraplength=700,
            justify="left",
            foreground=self.controller.colors['error'],
            style='AboutPageContent.TLabel'
        ).pack(fill=tk.X, padx=30, pady=(1, 10), anchor="w")

        # Back to Home Button
        button_frame = ttk.Frame(self.scrollable_frame, style='AboutPage.TFrame')
        button_frame.pack(fill=tk.X, padx=20, pady=(20, 30))

        ttk.Button(
            button_frame,
            text="← Back to Home",
            command=lambda: self.controller.show_frame("HomePage"),
            style='TButton'
        ).pack(side=tk.LEFT, ipady=5, ipadx=20)


if __name__ == "__main__":
    try:
        app = App()
        app.mainloop()
    except Exception as e:
        logging.critical(f"Application encountered a fatal error: {e}", exc_info=True)
        messagebox.showerror("Fatal Error", f"The application encountered a critical error and will close:\n{str(e)}\n\nPlease check 'crop_diagnosis.log' for more details.")

```
